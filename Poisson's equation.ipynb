# Poisson's Equation in Electrostatics

Poisson's equation is one of the central field equations of electrostatics. It expresses, in local differential form, how electric potential is generated by charge density. In Jackson-style electrodynamics, it is not merely a convenient equation to solve: it is the entry point to boundary value problems, uniqueness theorems, Green's functions, the method of images, and the systematic treatment of conductors and dielectric media. This notebook develops Poisson's equation from first principles, emphasizes its physical meaning, and supplements the theory with computational demonstrations designed to build geometric and analytical intuition.

## Electrostatic Foundations

In electrostatics, Maxwell's equations reduce to

$$
\nabla \cdot \mathbf{E} = \frac{\rho}{\varepsilon_0},
\qquad
\nabla \times \mathbf{E} = 0.
$$

The vanishing curl implies that the electric field is conservative, so there exists a scalar potential $\phi(\mathbf{r})$ such that

$$
\mathbf{E} = -\nabla \phi.
$$

Substituting this into Gauss's law gives

$$
\nabla \cdot (-\nabla \phi) = \frac{\rho}{\varepsilon_0}.
$$

Since $\nabla \cdot \nabla = \nabla^2$, the Laplacian, one obtains

$$
-\nabla^2 \phi = \frac{\rho}{\varepsilon_0},
$$

or equivalently,

$$
\nabla^2 \phi = -\frac{\rho}{\varepsilon_0}.
$$

This is **Poisson's equation**.

In regions where no charge is present, $\rho = 0$, and Poisson's equation reduces to

$$
\nabla^2 \phi = 0,
$$

which is **Laplace's equation**. Thus Laplace's equation is simply the source-free limit of Poisson's equation.

## Physical Meaning

Poisson's equation relates the **local curvature** of the electrostatic potential to the local charge density. In one dimension,

$$
\frac{d^2 \phi}{dx^2} = -\frac{\rho(x)}{\varepsilon_0},
$$

so the sign of the charge density determines the sign of the second derivative of the potential.

- If $\rho > 0$, then $\nabla^2 \phi < 0$: the potential is locally concave downward.
- If $\rho < 0$, then $\nabla^2 \phi > 0$: the potential is locally concave upward.
- If $\rho = 0$, then the potential is harmonic: its value is constrained entirely by the surrounding boundary data.

This is an important conceptual point. The potential is not determined only by nearby charge; it is determined globally by both the source distribution and the boundary conditions. Poisson's equation supplies the local differential relation, while the boundary conditions close the problem.

When $\nabla^2 \phi = 0$, the solution is called a **harmonic function**. Harmonic functions have strong rigidity properties: they satisfy mean-value relations, cannot possess interior maxima or minima unless constant, and are uniquely determined by suitable boundary data. These features are central in electrostatic boundary value theory.

## Mathematical Structure

### The Laplacian

In Cartesian coordinates,

$$
\nabla^2 \phi = \frac{\partial^2 \phi}{\partial x^2} + \frac{\partial^2 \phi}{\partial y^2} + \frac{\partial^2 \phi}{\partial z^2}.
$$

In cylindrical coordinates $(s,\varphi,z)$,

$$
\nabla^2 \phi = \frac{1}{s}\frac{\partial}{\partial s}\left(s\frac{\partial \phi}{\partial s}\right)
+ \frac{1}{s^2}\frac{\partial^2 \phi}{\partial \varphi^2}
+ \frac{\partial^2 \phi}{\partial z^2}.
$$

In spherical coordinates $(r,\theta,\phi)$,

$$
\nabla^2 \phi = \frac{1}{r^2}\frac{\partial}{\partial r}\left(r^2\frac{\partial \phi}{\partial r}\right)
+ \frac{1}{r^2\sin\theta}\frac{\partial}{\partial \theta}\left(\sin\theta\frac{\partial \phi}{\partial \theta}\right)
+ \frac{1}{r^2\sin^2\theta}\frac{\partial^2 \phi}{\partial \phi^2}.
$$

### Linearity

Poisson's equation is linear. If $\phi_1$ and $\phi_2$ solve

$$
\nabla^2 \phi_1 = -\frac{\rho_1}{\varepsilon_0},
\qquad
\nabla^2 \phi_2 = -\frac{\rho_2}{\varepsilon_0},
$$

then for constants $a$ and $b$,

$$
\nabla^2 (a\phi_1 + b\phi_2) = -\frac{a\rho_1 + b\rho_2}{\varepsilon_0}.
$$

This superposition principle is the differential expression of the linearity of electrostatics in vacuum.

### Boundary Conditions and Uniqueness

A differential equation alone does not determine a unique potential. Boundary conditions are essential.

- **Dirichlet boundary conditions** specify the potential on the boundary: $\phi|_{\partial V}$ is given.
- **Neumann boundary conditions** specify the normal derivative on the boundary: $\partial \phi/\partial n$ is given.

For Poisson's equation, appropriate boundary data lead to uniqueness theorems. In particular, the solution is uniquely determined by the charge density together with Dirichlet data. With Neumann data, the solution is unique up to an additive constant, provided consistency conditions are satisfied.

This is not merely mathematical bookkeeping. Physically, the same charge distribution can produce different potentials in different geometries because conductors, grounded boundaries, and asymptotic conditions alter the admissible solution.

## Connection with Coulomb's Law

For a point charge $q$ located at $\mathbf{r}_0$, the electrostatic potential is

$$
\phi(\mathbf{r}) = \frac{1}{4\pi\varepsilon_0}\frac{q}{|\mathbf{r}-\mathbf{r}_0|}.
$$

Away from the source point, this potential satisfies Laplace's equation. Indeed, for $\mathbf{r} \neq \mathbf{r}_0$,

$$
\nabla^2 \left(\frac{1}{|\mathbf{r}-\mathbf{r}_0|}\right)=0.
$$

However, globally the potential is singular at the source. The correct distributional identity is

$$
\nabla^2\left(\frac{1}{r}\right) = -4\pi\delta^{(3)}(\mathbf{r}).
$$

This statement means that for any sufficiently smooth test function $f(\mathbf{r})$,

$$
\iiint f(\mathbf{r})\,\nabla^2\left(\frac{1}{r}\right)d^3r = -4\pi f(0),
$$

after the singularity at the origin is treated correctly. Therefore,

$$
\nabla^2 \phi(\mathbf{r})
= \frac{q}{4\pi\varepsilon_0}\nabla^2\left(\frac{1}{|\mathbf{r}-\mathbf{r}_0|}\right)
= -\frac{q}{\varepsilon_0}\delta^{(3)}(\mathbf{r}-\mathbf{r}_0).
$$

Hence a point charge corresponds to the source density

$$
\rho(\mathbf{r}) = q\,\delta^{(3)}(\mathbf{r}-\mathbf{r}_0),
$$

and the Coulomb potential is precisely the Green-function solution of Poisson's equation for that source.

## Canonical Examples

### 1. Uniform Charge Density in a Slab

Suppose the potential depends only on $x$ and the region contains a constant charge density $\rho_0$. Then Poisson's equation becomes

$$
\frac{d^2\phi}{dx^2} = -\frac{\rho_0}{\varepsilon_0}.
$$

Integrating twice gives

$$
\phi(x) = -\frac{\rho_0}{2\varepsilon_0}x^2 + C_1 x + C_2.
$$

The constants are not determined by the differential equation itself; they are fixed by boundary conditions.

### 2. Point Charge in 3D

For a point charge,

$$
\rho(\mathbf{r}) = q\delta^{(3)}(\mathbf{r}-\mathbf{r}_0),
$$

and the corresponding solution is the Coulomb potential. This example shows how singular sources fit naturally within the Poisson framework.

### 3. Spherically Symmetric Charge Distribution

If $\rho = \rho(r)$ depends only on radius, then symmetry forces $\phi=\phi(r)$. Poisson's equation simplifies dramatically, allowing direct radial integration. This will be developed carefully in the next section.

### 4. Region with Zero Charge

Whenever $\rho=0$, Poisson's equation reduces to Laplace's equation. The potential is then harmonic and must be fixed entirely by boundary values or asymptotic behavior. Much of electrostatics consists of solving Laplace's equation subject to conductor boundary conditions.

### 5. A Simple Piecewise Example

Consider a one-dimensional interval in which charge is present only on part of the domain. In each charge-free region the potential satisfies Laplace's equation, hence it is linear. In the charged region it satisfies Poisson's equation, hence it is quadratic for constant density. The full solution is assembled by imposing continuity of $\phi$, continuity of $d\phi/dx$ when no surface charge is present, and the boundary conditions. This illustrates the basic logic of piecewise electrostatic construction.

## Spherical Symmetry

When $\phi=\phi(r)$ and $\rho=\rho(r)$, the angular derivatives vanish, so Poisson's equation becomes

$$
\frac{1}{r^2}\frac{d}{dr}\left(r^2\frac{d\phi}{dr}\right) = -\frac{\rho(r)}{\varepsilon_0}.
$$

This is the radial Poisson equation for spherical symmetry.

### Uniformly Charged Solid Sphere

Let a sphere of radius $R$ carry uniform volume charge density $\rho_0$. Then

$$
\rho(r)=
\begin{cases}
\rho_0, & r<R,\\
0, & r>R.
\end{cases}
$$

Inside the sphere, Poisson's equation reads

$$
\frac{1}{r^2}\frac{d}{dr}\left(r^2\frac{d\phi_{\text{in}}}{dr}\right) = -\frac{\rho_0}{\varepsilon_0}.
$$

Multiplying by $r^2$ and integrating once,

$$
r^2\frac{d\phi_{\text{in}}}{dr} = -\frac{\rho_0}{\varepsilon_0}\frac{r^3}{3} + C.
$$

Regularity at $r=0$ requires $C=0$, since otherwise $d\phi/dr \sim 1/r^2$ would diverge. Hence

$$
\frac{d\phi_{\text{in}}}{dr} = -\frac{\rho_0}{3\varepsilon_0}r.
$$

Integrating again,

$$
\phi_{\text{in}}(r)= -\frac{\rho_0}{6\varepsilon_0}r^2 + C_{\text{in}}.
$$

Outside the sphere, $\rho=0$, so

$$
\frac{1}{r^2}\frac{d}{dr}\left(r^2\frac{d\phi_{\text{out}}}{dr}\right)=0,
$$

whose spherically symmetric solution is

$$
\phi_{\text{out}}(r)=\frac{A}{r}+B.
$$

Imposing $\phi_{\text{out}}\to 0$ as $r\to\infty$ gives $B=0$. The total charge is

$$
Q = \frac{4}{3}\pi R^3\rho_0,
$$

so outside the sphere,

$$
\phi_{\text{out}}(r)=\frac{1}{4\pi\varepsilon_0}\frac{Q}{r}.
$$

Matching the potential continuously at $r=R$ yields

$$
\phi_{\text{in}}(r)=\frac{\rho_0}{6\varepsilon_0}(3R^2-r^2),
\qquad
\phi_{\text{out}}(r)=\frac{\rho_0 R^3}{3\varepsilon_0 r}.
$$

The electric field follows from $\mathbf{E}=-\nabla\phi$:

$$
\mathbf{E}(r)=
\begin{cases}
\dfrac{\rho_0 r}{3\varepsilon_0}\,\hat{\mathbf{r}}, & r<R,\\
\dfrac{\rho_0 R^3}{3\varepsilon_0 r^2}\,\hat{\mathbf{r}}, & r>R.
\end{cases}
$$

The interior field grows linearly with $r$, while the exterior field is exactly that of a point charge carrying the total charge $Q$.

## Boundary Value Problem Perspective

Poisson's equation should be viewed as a boundary value problem rather than simply as a differential identity. In practical electrostatics, one often knows some combination of

- the charge density $\rho(\mathbf{r})$ in a region,
- the potential on conducting surfaces,
- or the normal component of the electric field on a boundary.

Conductors are especially important because, in electrostatic equilibrium, they are equipotentials. Thus conductor surfaces naturally impose Dirichlet boundary conditions. Grounded conductors impose $\phi=0$.

The uniqueness theorems justify the whole enterprise of constructing solutions by symmetry arguments, separation of variables, image charges, or Green's functions. Once one finds a candidate solution satisfying Poisson's equation and the prescribed boundary conditions, the uniqueness theorem guarantees that it is the solution.

## Green's Function Preview

The free-space Green's function for the Laplacian is

$$
G(\mathbf{r},\mathbf{r}') = \frac{1}{|\mathbf{r}-\mathbf{r}'|},
$$

up to the conventional factor $1/(4\pi)$. Because

$$
\nabla^2\left(\frac{1}{|\mathbf{r}-\mathbf{r}'|}\right) = -4\pi\delta^{(3)}(\mathbf{r}-\mathbf{r}'),
$$

the formal free-space solution of Poisson's equation is

$$
\phi(\mathbf{r}) = \frac{1}{4\pi\varepsilon_0}\int \frac{\rho(\mathbf{r}')}{|\mathbf{r}-\mathbf{r}'|} \, d^3r'.
$$

This is the convolution of the source density with the free-space Green's function. It is the continuous superposition of Coulomb potentials due to infinitesimal charge elements. In bounded regions, the Green's function must be modified so that it also encodes the boundary conditions. That is the bridge to later notebooks on Green's functions, boundary value theory, and the method of images.

## Computational Section

The computational parts below are designed to reinforce the theory rather than replace it. We begin with one-dimensional Dirichlet problems, where analytical solutions are available for comparison, and then move to a simple two-dimensional finite-difference solution on a rectangular domain.

For visual clarity, the examples use dimensionless units with $\varepsilon_0=1$ unless stated otherwise. This changes the numerical scale but not the mathematical structure.

```python
import numpy as np
import matplotlib.pyplot as plt

# Clean notebook-oriented plotting style.
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 12
plt.rcParams['axes.titlesize'] = 13
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 10


def solve_poisson_1d_dirichlet(x, rho, phi_left=0.0, phi_right=0.0, epsilon0=1.0):
    """Solve phi'' = -rho/epsilon0 on a 1D grid with Dirichlet boundary conditions."""
    dx = x[1] - x[0]
    n = len(x)
    interior = n - 2

    # Standard second-difference matrix for the interior points.
    A = np.diag(-2.0 * np.ones(interior))
    A += np.diag(np.ones(interior - 1), k=1)
    A += np.diag(np.ones(interior - 1), k=-1)
    A /= dx**2

    b = -rho[1:-1] / epsilon0
    b[0] -= phi_left / dx**2
    b[-1] -= phi_right / dx**2

    phi_interior = np.linalg.solve(A, b)

    phi = np.zeros(n)
    phi[0] = phi_left
    phi[-1] = phi_right
    phi[1:-1] = phi_interior
    return phi


def electric_field_1d(x, phi):
    """Compute E = -dphi/dx with centered differences where possible."""
    return -np.gradient(phi, x)


def discrete_laplacian_1d(x, phi):
    """Compute the numerical second derivative of phi."""
    dx = x[1] - x[0]
    lap = np.zeros_like(phi)
    lap[1:-1] = (phi[2:] - 2 * phi[1:-1] + phi[:-2]) / dx**2
    return lap


def jacobi_poisson_2d(rho, dx, dy, boundary, epsilon0=1.0, iterations=2500):
    """Jacobi iteration for a 2D Poisson problem on a rectangular grid."""
    phi = boundary.copy()
    phi_new = phi.copy()
    inv_denom = 1.0 / (2.0 / dx**2 + 2.0 / dy**2)

    for _ in range(iterations):
        phi_new[1:-1, 1:-1] = (
            (phi[2:, 1:-1] + phi[:-2, 1:-1]) / dx**2
            + (phi[1:-1, 2:] + phi[1:-1, :-2]) / dy**2
            + rho[1:-1, 1:-1] / epsilon0
        ) * inv_denom

        # Reimpose the boundary values after each iteration.
        phi_new[0, :] = boundary[0, :]
        phi_new[-1, :] = boundary[-1, :]
        phi_new[:, 0] = boundary[:, 0]
        phi_new[:, -1] = boundary[:, -1]

        phi, phi_new = phi_new, phi

    return phi
```

### 1D Poisson Problem with Uniform Charge Density

Consider the interval $0 \le x \le L$ with constant charge density $\rho(x)=\rho_0$ and Dirichlet boundary conditions

$$
\phi(0)=0, \qquad \phi(L)=0.
$$

Then the analytical solution is

$$
\phi(x)=\frac{\rho_0}{2\varepsilon_0}x(L-x),
$$

which is quadratic, as expected from a constant source term. We now solve the same problem numerically using finite differences and compare the results.

```python
epsilon0 = 1.0
L = 1.0
rho0 = 2.0

x = np.linspace(0.0, L, 300)
rho_uniform = rho0 * np.ones_like(x)

# Numerical finite-difference solution.
phi_num = solve_poisson_1d_dirichlet(x, rho_uniform, phi_left=0.0, phi_right=0.0, epsilon0=epsilon0)

# Analytical solution for comparison.
phi_exact = (rho0 / (2 * epsilon0)) * x * (L - x)

fig, ax = plt.subplots()
ax.plot(x, phi_num, label='Numerical solution', linewidth=2)
ax.plot(x, phi_exact, '--', label='Analytical solution', linewidth=2)
ax.set_title('1D Poisson Equation with Uniform Charge Density')
ax.set_xlabel('$x$')
ax.set_ylabel(r'$\phi(x)$')
ax.legend()
plt.show()

max_error = np.max(np.abs(phi_num - phi_exact))
print(f'Maximum absolute error = {max_error:.3e}')
```

The numerical and analytical curves are essentially indistinguishable at this grid resolution. The quadratic shape reflects the constant source term, and the vanishing of the potential at both boundaries fixes the integration constants uniquely.

### Potential, Electric Field, and Curvature

For the same example, the electric field is

$$
E_x(x) = -\frac{d\phi}{dx},
$$

and the source term is related to the second derivative by

$$
\frac{d^2\phi}{dx^2} = -\frac{\rho_0}{\varepsilon_0}.
$$

This is the clearest way to visualize the physical meaning of Poisson's equation: the charge density controls the curvature of the potential.

```python
E_num = electric_field_1d(x, phi_num)
lap_num = discrete_laplacian_1d(x, phi_num)

fig, axes = plt.subplots(3, 1, figsize=(8, 11), sharex=True)

axes[0].plot(x, phi_num, color='tab:blue')
axes[0].set_ylabel(r'$\phi(x)$')
axes[0].set_title('Potential, Electric Field, and Numerical Curvature')

axes[1].plot(x, E_num, color='tab:green')
axes[1].set_ylabel(r'$E_x(x)$')

axes[2].plot(x, lap_num, label=r'$d^2\phi/dx^2$', color='tab:purple')
axes[2].axhline(-rho0 / epsilon0, linestyle='--', color='tab:red', label=r'$-\rho/\varepsilon_0$')
axes[2].set_ylabel(r'$\phi^{\prime\prime}(x)$')
axes[2].set_xlabel('$x$')
axes[2].legend()

plt.tight_layout()
plt.show()
```

The top panel shows the quadratic potential, the middle panel the linearly varying electric field, and the bottom panel the numerical second derivative. The agreement of $d^2\phi/dx^2$ with $-\rho/\varepsilon_0$ is a direct computational confirmation of Poisson's equation.

### Localized Charge Distribution in 1D

A more realistic source is a localized smooth charge distribution. Let

$$
\rho(x)=\rho_c\exp\left[-\frac{(x-x_0)^2}{\sigma^2}\right],
$$

on a finite interval with grounded boundaries. This is the one-dimensional analogue of a localized charge cloud. Solving Poisson's equation for such a source illustrates how a smooth source produces a smooth potential whose curvature is strongest where the charge is concentrated.

```python
x_local = np.linspace(0.0, 1.0, 500)
x0 = 0.35
sigma = 0.08
rho_c = 8.0

# Localized Gaussian source on a grounded interval.
rho_local = rho_c * np.exp(-((x_local - x0) / sigma) ** 2)
phi_local = solve_poisson_1d_dirichlet(x_local, rho_local, phi_left=0.0, phi_right=0.0, epsilon0=1.0)
E_local = electric_field_1d(x_local, phi_local)

fig, axes = plt.subplots(3, 1, figsize=(8, 11), sharex=True)

axes[0].plot(x_local, rho_local, color='tab:red')
axes[0].set_ylabel(r'$\rho(x)$')
axes[0].set_title('Localized Charge Distribution and Its Electrostatic Response')

axes[1].plot(x_local, phi_local, color='tab:blue')
axes[1].set_ylabel(r'$\phi(x)$')

axes[2].plot(x_local, E_local, color='tab:green')
axes[2].set_ylabel(r'$E_x(x)$')
axes[2].set_xlabel('$x$')

plt.tight_layout()
plt.show()
```

The source is concentrated near $x=x_0$, and the potential responds by developing its strongest curvature in the same region. The electric field varies most rapidly near the charge cloud and adjusts smoothly to satisfy the grounded boundary conditions at the ends of the interval.

### 2D Finite-Difference Poisson Problem on a Rectangular Domain

To move beyond one dimension, consider a rectangular domain with zero potential on the boundary and a localized interior charge distribution. The discrete Poisson equation on a grid is the natural finite-difference analogue of

$$
\nabla^2\phi(x,y) = -\rho(x,y).
$$

We solve it numerically by Jacobi iteration. The resulting heatmaps illustrate how the potential is shaped jointly by the source and the boundary.

```python
nx, ny = 81, 81
x2 = np.linspace(-1.0, 1.0, nx)
y2 = np.linspace(-1.0, 1.0, ny)
dx = x2[1] - x2[0]
dy = y2[1] - y2[0]
X2, Y2 = np.meshgrid(x2, y2, indexing='ij')

# Localized 2D source centered away from the origin.
rho_2d = 40.0 * np.exp(-(((X2 - 0.2) / 0.22) ** 2 + ((Y2 + 0.15) / 0.18) ** 2))

# Grounded boundary: phi = 0 on all edges.
boundary = np.zeros_like(rho_2d)
phi_2d = jacobi_poisson_2d(rho_2d, dx, dy, boundary, epsilon0=1.0, iterations=3000)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

im0 = axes[0].imshow(
    rho_2d.T,
    extent=[x2.min(), x2.max(), y2.min(), y2.max()],
    origin='lower',
    cmap='magma',
    aspect='equal'
)
axes[0].set_title('Source Density $\\rho(x,y)$')
axes[0].set_xlabel('$x$')
axes[0].set_ylabel('$y$')
fig.colorbar(im0, ax=axes[0], shrink=0.9)

im1 = axes[1].imshow(
    phi_2d.T,
    extent=[x2.min(), x2.max(), y2.min(), y2.max()],
    origin='lower',
    cmap='viridis',
    aspect='equal'
)
axes[1].set_title('Numerical Potential $\\phi(x,y)$')
axes[1].set_xlabel('$x$')
axes[1].set_ylabel('$y$')
fig.colorbar(im1, ax=axes[1], shrink=0.9)

plt.tight_layout()
plt.show()
```

The left panel shows the localized charge density, while the right panel shows the resulting potential constrained by grounded boundaries. The potential is largest near the source but is forced to vanish at the edges, so the geometry of the domain visibly shapes the solution. This is exactly the boundary-value viewpoint emphasized in advanced electrostatics.

### Electric Field from the 2D Numerical Solution

Once the potential is known, the electric field follows from

$$
\mathbf{E} = -\nabla \phi.
$$

A quiver plot provides a useful geometric picture of how field lines tend to point away from a positive source and toward the boundary configuration that enforces the potential constraints.

```python
# Compute the electric field from the numerical potential.
Ex_2d = -np.gradient(phi_2d, dx, axis=0)
Ey_2d = -np.gradient(phi_2d, dy, axis=1)

# Downsample for a readable vector field plot.
step = 4

fig, ax = plt.subplots(figsize=(7, 6))
ax.contourf(X2, Y2, phi_2d, levels=30, cmap='viridis')
ax.quiver(
    X2[::step, ::step],
    Y2[::step, ::step],
    Ex_2d[::step, ::step],
    Ey_2d[::step, ::step],
    color='white',
    pivot='mid',
    scale=120
)
ax.set_title('Electric Field Derived from the 2D Potential')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_aspect('equal')
plt.show()
```

The vector field is obtained directly from the gradient of the numerical potential. This reinforces the conceptual sequence

$$
\rho \longrightarrow \phi \longrightarrow \mathbf{E},
$$

which is the basic logic of electrostatic field determination in Poisson problems.

## Key Takeaways

- Poisson's equation, $\nabla^2\phi=-\rho/\varepsilon_0$, is the local differential statement that charge density generates electrostatic potential.
- Laplace's equation is the source-free case and governs harmonic potentials in regions without charge.
- The Laplacian measures curvature, so Poisson's equation relates local curvature of the potential to the local source density.
- Boundary conditions are indispensable: the same differential equation can admit different physical solutions in different geometries.
- The Coulomb potential is the Green-function solution for a point-charge source, with the singular structure encoded by the distributional identity $\nabla^2(1/r)=-4\pi\delta^{(3)}(\mathbf{r})$.
- Spherical symmetry reduces Poisson's equation to an ordinary differential equation, making exact solutions possible for important charge distributions.
- Numerical finite-difference methods provide direct intuition for how source distributions and boundaries shape the potential and electric field.

## Common Pitfalls

- Confusing Poisson's equation with Laplace's equation by forgetting whether the region actually contains charge.
- Treating the source term as sufficient by itself and neglecting the boundary conditions needed for uniqueness.
- Forgetting that $\phi$ must generally be continuous across interfaces, while derivatives may jump when singular surface sources are present.
- Using the Coulomb potential without recognizing that its Laplacian contains a delta distribution at the source point.
- Ignoring coordinate-system Jacobians when writing the Laplacian or delta function in curvilinear coordinates.
- Interpreting numerical solutions without checking whether the chosen boundary conditions match the intended physical problem.

## Connection to Later Topics

Poisson's equation is the foundation for several major themes in electrodynamics:

- **Green's functions:** free-space and boundary-adapted propagators provide systematic integral solutions.
- **Boundary value problems:** conductor geometries, grounded surfaces, and prescribed normal fields all enter through boundary conditions.
- **Method of images:** an elegant construction of solutions that exploits uniqueness.
- **Multipole expansion:** the large-distance structure of Poisson solutions is organized by moments of the source distribution.
- **Dielectrics and material media:** Poisson's equation generalizes to media through constitutive relations and polarization charge.
- **Electrodynamics more broadly:** the electrostatic potential problem provides intuition for later wave equations, retarded potentials, and field-theoretic Green-function methods.
