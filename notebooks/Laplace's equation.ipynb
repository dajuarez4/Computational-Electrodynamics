# Laplace's Equation in Electrostatics

Laplace's equation is the central differential equation of source-free electrostatics. Whenever one studies the electric potential in a region containing no volume charge, the problem reduces to finding a harmonic function subject to appropriate boundary conditions. This makes Laplace's equation foundational for conductor problems, separation of variables, uniqueness theorems, multipole expansions, and Green-function methods. In the spirit of Jackson, the emphasis here is on the interplay between mathematical structure and physical interpretation: the equation is simple in form, but rich in consequences.

## Electrostatic Foundations

In electrostatics, Maxwell's equations reduce to

$$
\nabla \cdot \mathbf{E} = \frac{\rho}{\varepsilon_0},
\qquad
\nabla \times \mathbf{E} = 0.
$$

Since the electric field is irrotational, it can be written in terms of a scalar potential:

$$
\mathbf{E} = -\nabla \phi.
$$

Substituting into Gauss's law yields

$$
\nabla \cdot (-\nabla \phi) = \frac{\rho}{\varepsilon_0},
$$

or

$$
\nabla^2 \phi = -\frac{\rho}{\varepsilon_0}.
$$

This is Poisson's equation. In a region where the charge density vanishes,

$$
\rho = 0,
$$

Poisson's equation reduces immediately to

$$
\nabla^2 \phi = 0.
$$

This is **Laplace's equation**. It governs the electrostatic potential in charge-free regions. The phrase "charge-free" must be understood locally: the region under study contains no volume charge, even though charges may exist elsewhere and influence the solution through boundary conditions.

## Physical Meaning

Laplace's equation states that the Laplacian of the potential vanishes:

$$
\nabla^2 \phi = 0.
$$

Physically, this means there is no local source term in the region. The potential is not being generated there by volume charge density. Instead, it is determined entirely by the surrounding boundaries and by sources located outside the region.

Because the source term vanishes, the potential is smooth in a strong sense. It cannot have the sharp local curvature associated with a nonzero charge density. In one dimension, Laplace's equation says simply

$$
\frac{d^2\phi}{dx^2}=0,
$$

so the potential has zero curvature and must be linear. In higher dimensions, the statement is subtler but similar in spirit: the potential has no local excess or deficit relative to its surroundings.

A useful heuristic is that in a source-free region the value of the potential at a point is not determined by local generation, but by the way the entire surrounding boundary constrains the field. This is why Laplace problems are fundamentally boundary value problems.

## Mathematical Form

In Cartesian coordinates,

$$
\nabla^2 \phi = \frac{\partial^2 \phi}{\partial x^2} + \frac{\partial^2 \phi}{\partial y^2} + \frac{\partial^2 \phi}{\partial z^2}.
$$

In cylindrical coordinates $(s,\varphi,z)$,

$$
\nabla^2 \phi = \frac{1}{s}\frac{\partial}{\partial s}\left(s\frac{\partial \phi}{\partial s}\right)
+ \frac{1}{s^2}\frac{\partial^2 \phi}{\partial \varphi^2}
+ \frac{\partial^2 \phi}{\partial z^2}.
$$

In spherical coordinates $(r,\theta,\phi)$,

$$
\nabla^2 \phi = \frac{1}{r^2}\frac{\partial}{\partial r}\left(r^2\frac{\partial \phi}{\partial r}\right)
+ \frac{1}{r^2\sin\theta}\frac{\partial}{\partial \theta}\left(\sin\theta\frac{\partial \phi}{\partial \theta}\right)
+ \frac{1}{r^2\sin^2\theta}\frac{\partial^2 \phi}{\partial \phi^2}.
$$

Laplace's equation is linear. If $\phi_1$ and $\phi_2$ are harmonic, then any linear combination $a\phi_1+b\phi_2$ is also harmonic. This superposition principle is essential both analytically and physically.

Solutions to Laplace's equation are called **harmonic functions**. Harmonicity is not merely terminology: it signals a special class of functions with strong rigidity, mean-value behavior, and uniqueness properties that are unusually powerful in electrodynamics.

## Fundamental Properties of Harmonic Functions

### Mean Value Property

A harmonic function has the property that its value at a point equals the average of its values over any sphere centered at that point, provided the sphere lies entirely within the source-free region. Symbolically,

$$
\phi(\mathbf{r}_0) = \frac{1}{4\pi a^2}\int_{|\mathbf{r}-\mathbf{r}_0|=a} \phi(\mathbf{r})\, dA.
$$

This expresses mathematically the idea that harmonic functions have no local source or sink. The value at the center is completely balanced by the surrounding values.

### Maximum and Minimum Principles

A nonconstant harmonic function cannot attain a true maximum or minimum at an interior point of a connected region. Any extremum must occur on the boundary.

This is one of the most important structural facts about Laplace's equation. If an interior maximum existed, then the surrounding average would have to be smaller than the central value, contradicting the mean value property unless the function were constant.

### Consequence: No Isolated Interior Extrema

A harmonic function cannot have isolated interior maxima or minima. Physically, this means the electrostatic potential in a source-free region cannot form a local well or hill by itself. The field configuration is imposed from the boundary, not spontaneously generated in the interior.

### Uniqueness

If two harmonic functions satisfy the same boundary conditions in a region, their difference is harmonic and vanishes on the boundary. By the maximum principle, that difference must vanish everywhere. Hence the solution is unique.

This is why boundary conditions completely determine the potential in a well-posed Laplace problem.

## Boundary Conditions

Laplace's equation is fundamentally a boundary value problem.

- **Dirichlet boundary conditions** specify the potential on the boundary: $\phi|_{\partial V}$ is given.
- **Neumann boundary conditions** specify the normal derivative: $\partial \phi / \partial n$ is given on the boundary.
- **Mixed boundary conditions** prescribe potential on part of the boundary and normal derivative on another part.

In electrostatics, Dirichlet conditions arise naturally for conducting surfaces held at fixed potential. Neumann conditions correspond to fixing the normal component of the electric field, since

$$
E_n = -\frac{\partial \phi}{\partial n}.
$$

Mixed problems appear when different physical surfaces impose different constraints. The mathematical character of the solution is inseparable from the physical meaning of the boundary.

## One-Dimensional Example

In one dimension, Laplace's equation is

$$
\frac{d^2 \phi}{dx^2} = 0.
$$

Integrating twice gives

$$
\phi(x)=Ax+B.
$$

Thus the potential is linear in any source-free interval. If one imposes boundary values

$$
\phi(0)=V_0, \qquad \phi(d)=V_1,
$$

the solution is

$$
\phi(x)=V_0 + \frac{V_1-V_0}{d}x.
$$

This is the potential between parallel plates with fixed potentials, neglecting fringing. The electric field is uniform:

$$
E_x = -\frac{d\phi}{dx} = -\frac{V_1-V_0}{d}.
$$

The absence of curvature reflects the absence of charge between the plates.

## Two-Dimensional Cartesian Solutions

In a rectangular geometry, one often solves Laplace's equation by separation of variables. Let

$$
\phi(x,y)=X(x)Y(y).
$$

Substituting into

$$
\frac{\partial^2 \phi}{\partial x^2} + \frac{\partial^2 \phi}{\partial y^2}=0,
$$

gives

$$
\frac{X''}{X} + \frac{Y''}{Y}=0.
$$

Since the first term depends only on $x$ and the second only on $y$, each must equal a constant. Writing

$$
\frac{X''}{X}=k^2,
\qquad
\frac{Y''}{Y}=-k^2,
$$

one obtains the ordinary differential equations

$$
X''-k^2X=0,
\qquad
Y''+k^2Y=0.
$$

Depending on the geometry and boundary conditions, one may choose the hyperbolic or trigonometric behavior to lie in either coordinate. The boundary conditions select the allowed separation constants and determine the expansion coefficients.

### Canonical Rectangular Example

Consider the rectangle $0<x<a$, $0<y<b$ with

$$
\phi(0,y)=0,
\quad
\phi(a,y)=0,
\quad
\phi(x,0)=0,
\quad
\phi(x,b)=V_0\sin\left(\frac{\pi x}{a}\right).
$$

The solution is

$$
\phi(x,y)=V_0\frac{\sinh\left(\frac{\pi y}{a}\right)}{\sinh\left(\frac{\pi b}{a}\right)}
\sin\left(\frac{\pi x}{a}\right).
$$

This is a classic example of a boundary-driven harmonic field: the entire interior potential is determined by the specified edge data.

## Spherical Symmetry and Angular Structure

In spherical coordinates, Laplace's equation is

$$
\frac{1}{r^2}\frac{\partial}{\partial r}\left(r^2\frac{\partial \phi}{\partial r}\right)
+ \frac{1}{r^2\sin\theta}\frac{\partial}{\partial \theta}\left(\sin\theta\frac{\partial \phi}{\partial \theta}\right)
+ \frac{1}{r^2\sin^2\theta}\frac{\partial^2 \phi}{\partial \phi^2}=0.
$$

For a spherically symmetric solution, $\phi=\phi(r)$, so the angular derivatives vanish and the equation reduces to

$$
\frac{1}{r^2}\frac{d}{dr}\left(r^2\frac{d\phi}{dr}\right)=0.
$$

Integrating once gives

$$
r^2\frac{d\phi}{dr}=C,
$$

and integrating again yields the general form

$$
\phi(r)=A+\frac{B}{r}.
$$

The constant term represents a uniform reference potential. The $1/r$ term is the source-free remnant of spherical geometry and is the form taken by the exterior potential of a localized charged object.

Regularity at the origin requires $B=0$, since $1/r$ diverges at $r=0$. By contrast, in an exterior region extending to infinity, physical decay often requires $A=0$ if one chooses the zero of potential at infinity.

More generally, separation in spherical coordinates leads to angular solutions described by Legendre polynomials and spherical harmonics. That structure underlies the multipole expansion, but here we restrict attention to the simplest radial sector.

## Cylindrical Symmetry

In cylindrical coordinates $(s,\varphi,z)$, Laplace's equation is

$$
\frac{1}{s}\frac{\partial}{\partial s}\left(s\frac{\partial \phi}{\partial s}\right)
+ \frac{1}{s^2}\frac{\partial^2 \phi}{\partial \varphi^2}
+ \frac{\partial^2 \phi}{\partial z^2}=0.
$$

If axial symmetry is present and the solution is independent of $\varphi$ and $z$, then the equation reduces to

$$
\frac{1}{s}\frac{d}{ds}\left(s\frac{d\phi}{ds}\right)=0.
$$

Integrating gives

$$
\phi(s)=A\ln s + B.
$$

Such forms arise naturally in problems involving long cylindrical conductors or coaxial geometries. The logarithm is the cylindrical analogue of the $1/r$ behavior familiar from spherical symmetry.

## Connection to Conductors

Inside a conductor in electrostatic equilibrium, the electric field vanishes:

$$
\mathbf{E}=0.
$$

Therefore the potential is constant throughout the conductor. In any charge-free conducting region, the potential satisfies Laplace's equation, but the physical equilibrium condition is even stronger: it is constant.

Outside a conductor, in the surrounding vacuum where no volume charge is present, the potential again satisfies Laplace's equation. The conductor surface then supplies the boundary condition. This is why conductor problems are the natural habitat of Laplace's equation. The geometry may be complicated, but once the boundary potential is known, the entire exterior field is in principle determined.

## Canonical Electrodynamics Examples

### 1. Potential Between Infinite Parallel Plates

Between two ideal infinite plates held at fixed potentials, the region between them is charge-free. Hence the potential satisfies Laplace's equation and is linear in the coordinate normal to the plates.

### 2. Spherically Symmetric Source-Free Region

If a spherical shell surrounds a region containing no charge, the potential inside that region satisfies Laplace's equation. Regularity at the origin often forces the solution to be constant unless nontrivial angular dependence is imposed from the boundary.

### 3. Exterior Region Outside a Charged Conductor

Outside a charged conductor there may be no volume charge in the vacuum region of interest, so the exterior potential satisfies Laplace's equation. The conductor surface and the asymptotic behavior determine the solution.

### 4. Simple Rectangular Boundary Value Problem

A rectangular region with prescribed edge potentials is a pure Laplace problem. The interior solution is generated entirely by the boundaries; there is no source term in the bulk.

Each of these is a Laplace problem rather than a Poisson problem because the governing region contains no volume charge density.

## Uniqueness Theorem

A central theorem states that the solution to Laplace's equation in a region is uniquely determined by suitable boundary conditions.

### Dirichlet Uniqueness

Suppose $\phi_1$ and $\phi_2$ are two solutions of Laplace's equation in the same region and take the same values on the boundary. Let

$$
u = \phi_1 - \phi_2.
$$

Then $u$ satisfies

$$
\nabla^2 u = 0,
$$

and $u=0$ on the boundary. By the maximum and minimum principles, a harmonic function whose boundary values vanish cannot achieve a positive maximum or negative minimum inside unless it is identically zero. Therefore $u=0$ everywhere, so

$$
\phi_1 = \phi_2.
$$

Hence the solution is unique.

### Physical Significance

This theorem is indispensable in electrodynamics. It justifies solution methods that appear indirect, such as the method of images. If one constructs any potential that satisfies Laplace's equation in the relevant region and obeys the correct boundary conditions, the uniqueness theorem guarantees that it is the physical solution.

## Computational Section

The numerical work below illustrates how Laplace's equation propagates boundary information into the interior. Since the equation contains no source term, the entire structure of the solution is controlled by geometry and boundary values.

For simplicity, we work in dimensionless units. The emphasis is on mathematical and physical structure rather than on units.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 12
plt.rcParams['axes.titlesize'] = 13
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 10


def solve_laplace_1d(x, phi_left, phi_right):
    """Exact 1D Laplace solution between two fixed boundaries."""
    return phi_left + (phi_right - phi_left) * (x - x[0]) / (x[-1] - x[0])


def jacobi_laplace_2d(boundary, iterations=3000):
    """Jacobi iteration for Laplace's equation on a rectangular grid."""
    phi = boundary.copy()
    phi_new = phi.copy()
    fixed = np.zeros_like(boundary, dtype=bool)
    fixed[0, :] = True
    fixed[-1, :] = True
    fixed[:, 0] = True
    fixed[:, -1] = True

    for _ in range(iterations):
        phi_new[1:-1, 1:-1] = 0.25 * (
            phi[2:, 1:-1] + phi[:-2, 1:-1] + phi[1:-1, 2:] + phi[1:-1, :-2]
        )
        phi_new[fixed] = boundary[fixed]
        phi, phi_new = phi_new, phi

    return phi


def analytic_rectangle_mode(x, y, a=1.0, b=1.0, V0=1.0):
    """Analytical separated solution for a single sinusoidal top boundary mode."""
    X, Y = np.meshgrid(x, y, indexing='ij')
    return V0 * np.sin(np.pi * X / a) * np.sinh(np.pi * Y / a) / np.sinh(np.pi * b / a)
```

### 1D Laplace Solution Between Two Boundaries

The simplest Laplace problem is the source-free interval with fixed endpoint potentials. Since the second derivative vanishes, the solution must be linear. Numerically and analytically there is nothing to distinguish: the equation itself already enforces the exact form.

This model corresponds to the electrostatic potential between ideal parallel plates when edge effects are neglected.

```python
x = np.linspace(0.0, 1.0, 300)
V0 = 5.0
V1 = -2.0

phi_1d = solve_laplace_1d(x, V0, V1)
E_1d = -np.gradient(phi_1d, x)

fig, axes = plt.subplots(2, 1, figsize=(8, 8), sharex=True)

axes[0].plot(x, phi_1d, color='tab:blue')
axes[0].set_ylabel(r'$\phi(x)$')
axes[0].set_title('One-Dimensional Laplace Solution')

axes[1].plot(x, E_1d, color='tab:green')
axes[1].set_ylabel(r'$E_x$')
axes[1].set_xlabel('$x$')

plt.tight_layout()
plt.show()

print(f'Uniform electric field = {E_1d.mean():.6f}')
```

The potential varies linearly between the two boundaries, and the electric field is constant. This is the simplest explicit illustration of the statement that source-free electrostatics in one dimension has zero curvature and therefore no local structure beyond that imposed by the boundaries.

### 2D Rectangular Problem: Analytical and Numerical Comparison

We now consider the canonical rectangular problem

$$
\phi(0,y)=0,
\quad
\phi(a,y)=0,
\quad
\phi(x,0)=0,
\quad
\phi(x,b)=V_0\sin\left(\frac{\pi x}{a}\right).
$$

For this problem the separated analytical solution is known exactly. We compare it with a finite-difference numerical solution to confirm both the mathematics and the computation.

```python
nx, ny = 81, 81
a, b = 1.0, 1.0
x2 = np.linspace(0.0, a, nx)
y2 = np.linspace(0.0, b, ny)
X2, Y2 = np.meshgrid(x2, y2, indexing='ij')
V_top = 1.0

# Dirichlet boundary data for the rectangle.
boundary = np.zeros((nx, ny))
boundary[:, -1] = V_top * np.sin(np.pi * x2 / a)

phi_num = jacobi_laplace_2d(boundary, iterations=4000)
phi_exact = analytic_rectangle_mode(x2, y2, a=a, b=b, V0=V_top)

fig, axes = plt.subplots(1, 3, figsize=(15, 4.8))

im0 = axes[0].imshow(phi_exact.T, extent=[0, a, 0, b], origin='lower', cmap='viridis', aspect='equal')
axes[0].set_title('Analytical Solution')
axes[0].set_xlabel('$x$')
axes[0].set_ylabel('$y$')
fig.colorbar(im0, ax=axes[0], shrink=0.85)

im1 = axes[1].imshow(phi_num.T, extent=[0, a, 0, b], origin='lower', cmap='viridis', aspect='equal')
axes[1].set_title('Numerical Solution')
axes[1].set_xlabel('$x$')
axes[1].set_ylabel('$y$')
fig.colorbar(im1, ax=axes[1], shrink=0.85)

im2 = axes[2].imshow((phi_num - phi_exact).T, extent=[0, a, 0, b], origin='lower', cmap='coolwarm', aspect='equal')
axes[2].set_title('Numerical Error')
axes[2].set_xlabel('$x$')
axes[2].set_ylabel('$y$')
fig.colorbar(im2, ax=axes[2], shrink=0.85)

plt.tight_layout()
plt.show()

print(f'Maximum absolute error = {np.max(np.abs(phi_num - phi_exact)):.3e}')
```

The numerical and analytical solutions agree closely. The interior potential is a smooth continuation of the sinusoidal boundary data on the top edge, attenuating toward the bottom grounded edge. This is a direct illustration of how boundary information propagates through a source-free region.

### Heatmap and Electric Field Visualization

Once the harmonic potential is known, the electric field follows from

$$
\mathbf{E}=-\nabla \phi.
$$

The vector field below shows how the field is shaped entirely by the chosen boundary values, since there is no source density in the interior.

```python
dx = x2[1] - x2[0]
dy = y2[1] - y2[0]
Ex = -np.gradient(phi_num, dx, axis=0)
Ey = -np.gradient(phi_num, dy, axis=1)

step = 4

fig, ax = plt.subplots(figsize=(7, 6))
cont = ax.contourf(X2, Y2, phi_num, levels=30, cmap='viridis')
ax.quiver(
    X2[::step, ::step],
    Y2[::step, ::step],
    Ex[::step, ::step],
    Ey[::step, ::step],
    color='white',
    pivot='mid',
    scale=20
)
ax.set_title('Electric Field from the Numerical Harmonic Potential')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_aspect('equal')
fig.colorbar(cont, ax=ax, shrink=0.85, label=r'$\phi$')
plt.show()
```

The field vectors point from higher to lower potential, as expected. Their structure is entirely boundary-driven: the interior contains no charge, so the field is the smooth adjustment required by the edge potentials.

### Changing Boundary Conditions Changes the Entire Solution

A defining feature of Laplace's equation is that changing the boundary data changes the solution everywhere, not merely near the altered boundary. We illustrate this by comparing two rectangular problems with different top-edge potentials while keeping the other boundaries grounded.

```python
boundary_case_1 = np.zeros((nx, ny))
boundary_case_1[:, -1] = np.sin(np.pi * x2)

boundary_case_2 = np.zeros((nx, ny))
boundary_case_2[:, -1] = np.sin(np.pi * x2) + 0.4 * np.sin(3 * np.pi * x2)

phi_case_1 = jacobi_laplace_2d(boundary_case_1, iterations=4000)
phi_case_2 = jacobi_laplace_2d(boundary_case_2, iterations=4000)

fig, axes = plt.subplots(1, 3, figsize=(15, 4.8))

im0 = axes[0].imshow(phi_case_1.T, extent=[0, a, 0, b], origin='lower', cmap='viridis', aspect='equal')
axes[0].set_title('Case 1: Single Boundary Mode')
axes[0].set_xlabel('$x$')
axes[0].set_ylabel('$y$')
fig.colorbar(im0, ax=axes[0], shrink=0.85)

im1 = axes[1].imshow(phi_case_2.T, extent=[0, a, 0, b], origin='lower', cmap='viridis', aspect='equal')
axes[1].set_title('Case 2: Modified Boundary Data')
axes[1].set_xlabel('$x$')
axes[1].set_ylabel('$y$')
fig.colorbar(im1, ax=axes[1], shrink=0.85)

im2 = axes[2].imshow((phi_case_2 - phi_case_1).T, extent=[0, a, 0, b], origin='lower', cmap='coolwarm', aspect='equal')
axes[2].set_title('Difference Between Solutions')
axes[2].set_xlabel('$x$')
axes[2].set_ylabel('$y$')
fig.colorbar(im2, ax=axes[2], shrink=0.85)

plt.tight_layout()
plt.show()
```

The difference plot shows that a change imposed on one boundary affects the entire interior solution. This global sensitivity is characteristic of elliptic equations and is one reason boundary conditions are so central in electrostatics.

## Key Takeaways

- Laplace's equation, $\nabla^2\phi=0$, governs the electrostatic potential in charge-free regions.
- Harmonic functions are linear under superposition and possess powerful structural properties such as the mean value property and the maximum principle.
- In source-free electrostatics, the potential is determined entirely by boundary conditions.
- The absence of interior extrema for nonconstant harmonic functions has direct physical meaning: there are no local sources or sinks in the region.
- One-dimensional Laplace problems produce linear potentials and uniform electric fields.
- Separation of variables in rectangular, cylindrical, and spherical coordinates is one of the main analytical tools for solving Laplace problems.
- Uniqueness theorems justify many classical techniques in electrodynamics, including image methods and boundary construction arguments.

## Common Pitfalls

- Forgetting that Laplace's equation applies only in regions where the volume charge density vanishes.
- Treating the solution as local in character and overlooking the fact that boundary conditions determine the entire interior field.
- Ignoring regularity conditions at the origin or asymptotic decay conditions at infinity.
- Confusing the absence of charge in a region with the absence of field in that region.
- Failing to distinguish between Dirichlet and Neumann data and the different physical meanings they carry.
- Using a candidate solution without checking that it satisfies both the differential equation and the boundary conditions.

## Connection to Later Topics

Laplace's equation leads directly into several major parts of electrodynamics:

- **Poisson's equation:** the natural extension when source terms are present.
- **Green's functions:** systematic construction of solutions adapted to boundaries.
- **Method of images:** a boundary-value technique justified by uniqueness.
- **Separation of variables:** the standard machinery for rectangular, cylindrical, and spherical geometries.
- **Multipole expansion:** the spherical-harmonic structure of source-free solutions at large distance.
- **Boundary value problems more broadly:** conductor geometries, cavities, waveguides, and later electromagnetic field problems all inherit the same elliptic logic.
