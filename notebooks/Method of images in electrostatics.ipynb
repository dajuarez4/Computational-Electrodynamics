# The Method of Images in Electrostatics

The method of images is one of the most elegant devices in electrostatics for solving boundary value problems that would otherwise be awkward to attack directly. Its power does not come from a mysterious physical replacement of conductors by charges, but from a rigorous mathematical fact: if one constructs a potential that satisfies Poisson's or Laplace's equation in the physical region, obeys the required boundary conditions, and has the correct behavior at infinity, then the uniqueness theorem guarantees that this potential is the physical solution. In favorable geometries, a carefully chosen set of fictitious image charges enforces those conditions exactly.

## Motivation and Conceptual Foundation

In electrostatics the potential satisfies

$$
\nabla^2 \phi(\mathbf{r}) = -\frac{\rho(\mathbf{r})}{\varepsilon_0}
$$

in regions containing charge, and Laplace's equation

$$
\nabla^2 \phi(\mathbf{r}) = 0
$$

in charge-free regions. The formal statement is simple; the difficulty lies in the boundary conditions. Conducting surfaces impose strong constraints on the admissible solution. For a perfect conductor in electrostatic equilibrium, the surface is an equipotential, and for a grounded conductor one has the Dirichlet condition $\phi=0$ on the surface.

Direct solution of Poisson's or Laplace's equation with nontrivial boundaries is generally difficult. The method of images replaces part of the geometric complexity by an auxiliary charge configuration chosen so that the resulting potential automatically satisfies the required boundary condition on the conductor. The image charges are not physical charges in the laboratory region. They are mathematical devices placed in an auxiliary region, usually the part of space occupied by the conductor or otherwise excluded from the physical domain.

The decisive point is therefore not whether the image construction looks plausible, but whether it yields a function $\phi$ that solves the correct differential equation in the physical region and satisfies the prescribed boundary data there. Once this is achieved, uniqueness elevates the construction from an ansatz to a proof.

## Why the Method Works

A convenient form of the electrostatic uniqueness theorem is the following.

**Uniqueness theorem.** Let $V$ be a region with boundary $S$. If the charge density $\rho(\mathbf{r})$ is specified in $V$ and either the potential $\phi$ is specified on $S$ (Dirichlet problem) or the normal derivative is specified together with appropriate compatibility conditions (Neumann problem), then the electrostatic solution is unique up to the usual additive constant in the pure Neumann case.

For the present notebook, the relevant case is predominantly Dirichlet. Suppose we construct a candidate potential $\phi_{\mathrm{cand}}$ such that:

1. $\nabla^2 \phi_{\mathrm{cand}} = -\rho/\varepsilon_0$ in the physical region.
2. $\phi_{\mathrm{cand}}$ satisfies the conductor boundary conditions.
3. $\phi_{\mathrm{cand}}$ has the appropriate asymptotic behavior as $|\mathbf{r}|\to\infty$ when the region is unbounded.

Then $\phi_{\mathrm{cand}}$ must be the physical potential.

This explains the central logic of the method of images. The image charges need not correspond to actual sources in the physical region. They need only be placed outside that region so that inside the physical domain the candidate potential has the correct singularities from the real charges alone and is otherwise harmonic. The conductor is then encoded indirectly through the boundary condition that the image construction enforces.

The method is therefore a special-purpose technique for constructing exact Dirichlet Green's functions in geometries with sufficient symmetry.

## Infinite Grounded Conducting Plane

Consider a point charge $q$ located at

$$
\mathbf{r}_0=(0,0,a), \qquad a>0,
$$

above the grounded conducting plane $z=0$. The physical region is the half-space $z>0$. We seek a potential satisfying

$$
\nabla^2\phi(\mathbf{r})=-\frac{q\,\delta(x)\delta(y)\delta(z-a)}{\varepsilon_0},
\qquad z>0,
$$

with boundary condition

$$
\phi(x,y,0)=0,
$$

and $\phi\to 0$ as $r\to\infty$.

The image construction is suggested by mirror symmetry. Place an image charge $-q$ at

$$
\mathbf{r}_0'=(0,0,-a).
$$

Then in the physical region the candidate potential is

$$
\phi(\mathbf{r})
= \frac{1}{4\pi\varepsilon_0}
\left[
\frac{q}{|\mathbf{r}-\mathbf{r}_0|}
-\frac{q}{|\mathbf{r}-\mathbf{r}_0'|}
\right].
$$

On the plane $z=0$ one has $|\mathbf{r}-\mathbf{r}_0|=|\mathbf{r}-\mathbf{r}_0'|$, so the two terms cancel identically and therefore $\phi=0$ on the boundary. Since the only singularity in $z>0$ is the real charge at $\mathbf{r}_0$, the candidate satisfies Poisson's equation in the half-space. By uniqueness, this is the physical solution.

In cylindrical coordinates $(\rho,\varphi,z)$ the potential becomes

$$
\phi(\rho,z)=\frac{q}{4\pi\varepsilon_0}
\left[
\frac{1}{\sqrt{\rho^2+(z-a)^2}}
-\frac{1}{\sqrt{\rho^2+(z+a)^2}}
\right], \qquad z>0.
$$

The electric field is obtained from $\mathbf{E}=-\nabla\phi$. Explicitly,

$$
\mathbf{E}(\mathbf{r})=
\frac{q}{4\pi\varepsilon_0}
\left[
\frac{\mathbf{r}-\mathbf{r}_0}{|\mathbf{r}-\mathbf{r}_0|^3}
-\frac{\mathbf{r}-\mathbf{r}_0'}{|\mathbf{r}-\mathbf{r}_0'|^3}
\right].
$$

### Induced Surface Charge, Total Charge, and Force

The induced surface charge density follows from the discontinuity of the normal field. Since the conductor occupies $z<0$, the outward normal from the conductor into the physical region is $\hat{\mathbf{z}}$, and

$$
\sigma(\rho)=\varepsilon_0 E_z(\rho,0^+)
=-\varepsilon_0\left.\frac{\partial \phi}{\partial z}\right|_{z=0^+}.
$$

Differentiating the image potential gives

$$
\sigma(\rho)
=-\frac{qa}{2\pi\left(\rho^2+a^2\right)^{3/2}}.
$$

The charge density is negative everywhere, as expected for a positive charge above a grounded conductor.

The total induced charge is

$$
Q_{\mathrm{ind}}=\int_0^{\infty}\sigma(\rho)\,2\pi\rho\,d\rho
=-qa\int_0^{\infty}\frac{\rho\,d\rho}{(\rho^2+a^2)^{3/2}}=-q.
$$

Thus the grounded plane carries precisely the negative of the external charge. This should not be confused with a statement about a finite isolated conductor; it is a consequence of the infinite grounded geometry.

The force on the real charge can be computed by replacing the conductor with the image charge and evaluating the Coulomb force exerted by the image on the real charge:

$$
\mathbf{F}=
\frac{1}{4\pi\varepsilon_0}
\frac{q(-q)}{(2a)^2}\,\hat{\mathbf{z}}_{\text{toward image}}
=-\frac{q^2}{16\pi\varepsilon_0 a^2}\,\hat{\mathbf{z}}.
$$

The force is attractive and scales as though the real charge interacted with a mirror charge a distance $2a$ away. Physically, the conductor polarizes so that negative surface charge accumulates beneath the point charge, drawing it toward the plane.

## Computational Demonstration: Grounded Plane

For visualization it is convenient to work in normalized units with $1/(4\pi\varepsilon_0)=1$, $q=1$, and $a=1$. This leaves the geometry unchanged and suppresses irrelevant unit factors. The code below plots equipotentials in the half-space, the electric field, the image geometry, and the induced surface charge density. It also checks numerically that the integrated induced charge approaches $-q$.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

plt.rcParams.update({
    "figure.figsize": (12, 9),
    "font.size": 11,
    "axes.grid": False,
})

k_c = 1.0  # normalized Coulomb constant: 1/(4 pi epsilon_0) = 1


def plane_potential(x, z, q=1.0, a=1.0):
    r_real = np.sqrt(x**2 + (z - a)**2)
    r_image = np.sqrt(x**2 + (z + a)**2)
    return k_c * q * (1.0 / r_real - 1.0 / r_image)


def plane_field(x, z, q=1.0, a=1.0):
    r_real = (x**2 + (z - a)**2) ** 1.5
    r_image = (x**2 + (z + a)**2) ** 1.5
    ex = k_c * q * (x / r_real - x / r_image)
    ez = k_c * q * ((z - a) / r_real - (z + a) / r_image)
    return ex, ez


def plane_sigma(rho, q=1.0, a=1.0):
    return -(q * a) / (2.0 * np.pi * (rho**2 + a**2) ** 1.5)


# Grid for potential and field in the physical region z > 0.
a = 1.0
q = 1.0
x = np.linspace(-4.0, 4.0, 401)
z = np.linspace(0.02, 4.0, 321)
X, Z = np.meshgrid(x, z)
phi_plane = plane_potential(X, Z, q=q, a=a)

# Mask the immediate neighborhood of the real charge to avoid plotting the singularity.
mask_charge = np.sqrt(X**2 + (Z - a)**2) < 0.12
phi_plane = np.ma.array(phi_plane, mask=mask_charge)

xq = np.linspace(-4.0, 4.0, 33)
zq = np.linspace(0.2, 3.8, 27)
Xq, Zq = np.meshgrid(xq, zq)
Exq, Ezq = plane_field(Xq, Zq, q=q, a=a)
mask_quiver = np.sqrt(Xq**2 + (Zq - a)**2) < 0.25
Exq = np.ma.array(Exq, mask=mask_quiver)
Ezq = np.ma.array(Ezq, mask=mask_quiver)

rho_plot = np.linspace(0.0, 8.0, 1200)
rho_int = np.linspace(0.0, 200.0, 40000)
sigma_vals = plane_sigma(rho_plot, q=q, a=a)
sigma_int = plane_sigma(rho_int, q=q, a=a)
q_induced_numeric = np.trapz(2.0 * np.pi * rho_int * sigma_int, rho_int)
force_plane = -q**2 / (4.0 * (2.0 * a) ** 2)

fig, axes = plt.subplots(2, 2, figsize=(13, 10))

levels = np.linspace(0.05, 2.0, 18)
axes[0, 0].contour(X, Z, phi_plane, levels=levels, cmap="viridis")
axes[0, 0].axhline(0.0, color="black", linewidth=2)
axes[0, 0].scatter([0.0], [a], color="crimson", s=70, label="real charge q")
axes[0, 0].set_title("Potential contours above the grounded plane")
axes[0, 0].set_xlabel("x / a")
axes[0, 0].set_ylabel("z / a")
axes[0, 0].set_xlim(-4, 4)
axes[0, 0].set_ylim(0, 4)
axes[0, 0].legend(loc="upper right")

axes[0, 1].contour(X, Z, phi_plane, levels=levels[::2], colors="0.7", linewidths=0.8)
axes[0, 1].quiver(Xq, Zq, Exq, Ezq, pivot="mid", color="tab:blue", scale=45)
axes[0, 1].axhline(0.0, color="black", linewidth=2)
axes[0, 1].scatter([0.0], [a], color="crimson", s=70)
axes[0, 1].set_title("Electric field in the physical region")
axes[0, 1].set_xlabel("x / a")
axes[0, 1].set_ylabel("z / a")
axes[0, 1].set_xlim(-4, 4)
axes[0, 1].set_ylim(0, 4)

axes[1, 0].plot(rho_plot, sigma_vals, color="tab:green", linewidth=2)
axes[1, 0].axhline(0.0, color="black", linewidth=1)
axes[1, 0].set_title("Induced surface charge density on the plane")
axes[1, 0].set_xlabel(r"$\rho / a$")
axes[1, 0].set_ylabel(r"$\sigma$ in normalized units")
axes[1, 0].set_xlim(0, 8)

axes[1, 1].axhline(0.0, color="black", linewidth=2, label="grounded plane")
axes[1, 1].scatter([0.0], [a], color="crimson", s=70, label="real charge q")
axes[1, 1].scatter([0.0], [-a], color="navy", s=70, label="image charge -q")
axes[1, 1].plot([0, 0], [-a, a], color="0.5", linestyle="--")
axes[1, 1].set_title("Image construction")
axes[1, 1].set_xlabel("x / a")
axes[1, 1].set_ylabel("z / a")
axes[1, 1].set_xlim(-2.5, 2.5)
axes[1, 1].set_ylim(-2.5, 2.5)
axes[1, 1].set_aspect("equal", adjustable="box")
axes[1, 1].legend(loc="upper right")

plt.tight_layout()
plt.show()

print(f"Numerical induced charge on the plane: {q_induced_numeric:.6f}")
print(f"Exact induced charge: {-q:.6f}")
print(f"Normalized force on the real charge: {force_plane:.6f} z-hat")
```

The contour plot makes the Dirichlet condition visually explicit: the conducting plane is the zero-potential boundary for the half-space problem. The electric field is everywhere normal to the plane, as required in electrostatics. The graph of $\sigma(\rho)$ shows the strongest induced charge directly beneath the point charge and the algebraic decay $\sigma\sim -\rho^{-3}$ at large radius. The numerical integral reproduces $Q_{\mathrm{ind}}=-q$, which is a useful computational check on the analytic derivation.

## Grounded Conducting Sphere with an External Point Charge

Now consider a grounded conducting sphere of radius $R$ centered at the origin and a point charge $q$ located on the positive $z$-axis at distance $a>R$ from the center. The physical region is the exterior $r>R$. The problem is to solve

$$
\nabla^2\phi(\mathbf{r})=-\frac{q\,\delta^{(3)}(\mathbf{r}-a\hat{\mathbf{z}})}{\varepsilon_0},
\qquad r>R,
$$

subject to

$$
\phi(R,\theta)=0, \qquad \phi(\mathbf{r})\to 0 \text{ as } r\to\infty.
$$

Unlike the plane, mirror reflection is no longer sufficient. The required image charge has both a different magnitude and a different location. One seeks an image charge $q'$ placed on the same axis at distance $b<R$ from the center. The candidate potential is

$$
\phi(\mathbf{r})=
\frac{1}{4\pi\varepsilon_0}
\left[
\frac{q}{|\mathbf{r}-a\hat{\mathbf{z}}|}
+
\frac{q'}{|\mathbf{r}-b\hat{\mathbf{z}}|}
\right],
\qquad r>R.
$$

To determine $q'$ and $b$, impose $\phi(R,\theta)=0$ for all polar angles $\theta$. On the sphere,

$$
|R\hat{\mathbf{r}}-a\hat{\mathbf{z}}|=\sqrt{R^2+a^2-2aR\cos\theta},
$$

and

$$
|R\hat{\mathbf{r}}-b\hat{\mathbf{z}}|=\sqrt{R^2+b^2-2bR\cos\theta}.
$$

The construction succeeds if these two denominators are proportional as functions of $\theta$. This occurs when

$$
b=\frac{R^2}{a},
$$

for then

$$
R^2+b^2-2bR\cos\theta
=\frac{R^2}{a^2}\left(a^2+R^2-2aR\cos\theta\right),
$$

so that

$$
|R\hat{\mathbf{r}}-b\hat{\mathbf{z}}|=\frac{R}{a}|R\hat{\mathbf{r}}-a\hat{\mathbf{z}}|.
$$

Therefore the boundary condition demands

$$
q'=-q\frac{R}{a}.
$$

The exterior potential is thus

$$
\phi(\mathbf{r})=
\frac{1}{4\pi\varepsilon_0}
\left[
\frac{q}{|\mathbf{r}-a\hat{\mathbf{z}}|}
-
\frac{qR/a}{|\mathbf{r}-(R^2/a)\hat{\mathbf{z}}|}
\right],
\qquad r>R.
$$

The image construction is less obvious than for the plane because the sphere does not arise from a Euclidean reflection. Instead, the geometry is governed by an inversion relation characteristic of spherical boundaries.

### Verification on the Surface, Induced Charge, and Force

On the spherical surface $r=R$, the proportionality above gives

$$
\phi(R,\theta)=\frac{1}{4\pi\varepsilon_0}
\left[
\frac{q}{|R\hat{\mathbf{r}}-a\hat{\mathbf{z}}|}
-
\frac{qR/a}{(R/a)|R\hat{\mathbf{r}}-a\hat{\mathbf{z}}|}
\right]=0.
$$

Hence the candidate satisfies the grounded boundary condition pointwise, and uniqueness identifies it as the physical solution in the exterior region.

The induced surface charge density follows from the outward radial field:

$$
\sigma(\theta)=-\varepsilon_0\left.\frac{\partial \phi}{\partial r}\right|_{r=R^+}.
$$

Carrying out the differentiation and simplifying gives

$$
\sigma(\theta)=
-\frac{q}{4\pi R}
\frac{a^2-R^2}{\left(a^2+R^2-2aR\cos\theta\right)^{3/2}}.
$$

This is everywhere negative for $a>R$ and is most concentrated near $\theta=0$, the point of closest approach to the external charge.

The total induced charge is

$$
Q_{\mathrm{ind}}=\int \sigma\,dA = 2\pi R^2\int_0^{\pi}\sigma(\theta)\sin\theta\,d\theta = q'=-q\frac{R}{a}.
$$

The total induced charge is therefore equal to the image charge, which is a general and useful fact for grounded conductors in image constructions of this type.

The force on the external charge is the Coulomb force due to the image charge:

$$
\mathbf{F}
=\frac{1}{4\pi\varepsilon_0}
\frac{qq'}{(a-b)^2}\,\hat{\mathbf{z}}
=-\frac{1}{4\pi\varepsilon_0}
\frac{q^2Ra}{(a^2-R^2)^2}\,\hat{\mathbf{z}}.
$$

Again the force is attractive. In the limiting regime $a=R+d$ with $d\ll R$ and $R\to\infty$, one finds

$$
\mathbf{F}\to -\frac{q^2}{16\pi\varepsilon_0 d^2}\,\hat{\mathbf{z}},
$$

which matches the plane result. Locally, a large sphere appears flat near the point closest to the charge.

## Comparison Between Plane and Sphere

The plane problem is governed by literal mirror symmetry: the image charge has equal magnitude and opposite sign, placed at the reflected point across the boundary. The sphere problem has no such Euclidean reflection. Instead, the correct image point is determined by spherical inversion, and the image magnitude is reduced by the geometric factor $R/a$.

The difference reflects the underlying symmetry of the boundary. A plane is invariant under translations parallel to itself and under reflection across itself. A sphere is invariant under rotations about its center, and the relevant transformation that preserves harmonic structure is inversion rather than translation. Nevertheless, the logic of the two problems is identical. In each case one constructs a potential that satisfies the field equation in the physical region, enforces the conductor boundary condition, and decays appropriately. The uniqueness theorem then does the essential conceptual work.

## Relation to Green's Functions

The method of images may be viewed as an explicit construction of a Dirichlet Green's function. For the grounded plane bounding the half-space $z>0$, the Dirichlet Green's function is

$$
G_D(\mathbf{r},\mathbf{r}')=
\frac{1}{|\mathbf{r}-\mathbf{r}'|}
-
\frac{1}{|\mathbf{r}-\mathbf{r}'^{\,*}|},
$$

where $\mathbf{r}'^{\,*}$ is the reflected source point across the plane. It vanishes when the field point lies on the boundary and reproduces the image solution immediately.

For the grounded sphere, the corresponding Green's function can also be written in image form:

$$
G_D(\mathbf{r},\mathbf{r}')=
\frac{1}{|\mathbf{r}-\mathbf{r}'|}
-
\frac{R}{r'}\frac{1}{\left|\mathbf{r}-\frac{R^2}{r'^2}\mathbf{r}'\right|},
\qquad r,r'>R.
$$

This notebook does not pursue the full Green's-function formalism, but the conceptual point is important: image charges are best understood as a compact way of encoding boundary conditions into a kernel. The method works only when such a kernel can be represented by a simple auxiliary source configuration.

## Canonical Derivations and Worked Results

For convenient reference, the principal formulas derived above are collected here.

### 1. Point charge near a grounded infinite plane

For a charge $q$ at height $a$ above the plane $z=0$,

$$
\phi(\rho,z)=\frac{q}{4\pi\varepsilon_0}
\left[
\frac{1}{\sqrt{\rho^2+(z-a)^2}}-
\frac{1}{\sqrt{\rho^2+(z+a)^2}}
\right], \qquad z>0.
$$

The induced surface charge density is

$$
\sigma(\rho)=-\frac{qa}{2\pi(\rho^2+a^2)^{3/2}},
$$

the total induced charge is

$$
Q_{\mathrm{ind}}=-q,
$$

and the force on the real charge is

$$
\mathbf{F}=-\frac{q^2}{16\pi\varepsilon_0 a^2}\,\hat{\mathbf{z}}.
$$

### 2. Point charge outside a grounded sphere

For a charge $q$ at $a\hat{\mathbf{z}}$ with $a>R$,

$$
q'=-q\frac{R}{a},
\qquad
b=\frac{R^2}{a}.
$$

The exterior potential is

$$
\phi(\mathbf{r})=
\frac{1}{4\pi\varepsilon_0}
\left[
\frac{q}{|\mathbf{r}-a\hat{\mathbf{z}}|}
+
\frac{q'}{|\mathbf{r}-b\hat{\mathbf{z}}|}
\right], \qquad r>R.
$$

The induced charge density is

$$
\sigma(\theta)=
-\frac{q}{4\pi R}
\frac{a^2-R^2}{\left(a^2+R^2-2aR\cos\theta\right)^{3/2}},
$$

the total induced charge is

$$
Q_{\mathrm{ind}}=q'=-q\frac{R}{a},
$$

and the force on the real charge is

$$
\mathbf{F}=-\frac{1}{4\pi\varepsilon_0}\frac{q^2Ra}{(a^2-R^2)^2}\,\hat{\mathbf{z}}.
$$

### 3. Physical interpretation

The image charges are not present in the physical region. They merely reproduce the same potential there as the conductor would produce. Their purpose is to satisfy the boundary condition indirectly while preserving the correct singularity structure from the real sources.

## Computational Demonstration: Grounded Sphere

We again use normalized units with $1/(4\pi\varepsilon_0)=1$. The following code visualizes the potential outside the grounded sphere, the electric field, the image construction, the induced surface charge density, and the effect of the conductor on the potential compared with free space. It also checks numerically that the boundary is equipotential and that the total induced charge matches the image charge.

```python
def sphere_image_parameters(q=1.0, R=1.0, a=2.0):
    q_image = -q * R / a
    b = R**2 / a
    return q_image, b


def sphere_potential_cartesian(x, z, q=1.0, R=1.0, a=2.0):
    q_image, b = sphere_image_parameters(q=q, R=R, a=a)
    r_real = np.sqrt(x**2 + (z - a)**2)
    r_image = np.sqrt(x**2 + (z - b)**2)
    return k_c * (q / r_real + q_image / r_image)


def sphere_field_cartesian(x, z, q=1.0, R=1.0, a=2.0):
    q_image, b = sphere_image_parameters(q=q, R=R, a=a)
    r_real = (x**2 + (z - a)**2) ** 1.5
    r_image = (x**2 + (z - b)**2) ** 1.5
    ex = k_c * (q * x / r_real + q_image * x / r_image)
    ez = k_c * (q * (z - a) / r_real + q_image * (z - b) / r_image)
    return ex, ez


def sphere_sigma(theta, q=1.0, R=1.0, a=2.0):
    denominator = (a**2 + R**2 - 2.0 * a * R * np.cos(theta)) ** 1.5
    return -(q / (4.0 * np.pi * R)) * (a**2 - R**2) / denominator


R = 1.0
a = 2.0
q = 1.0
q_image, b = sphere_image_parameters(q=q, R=R, a=a)

x = np.linspace(-3.0, 3.0, 451)
z = np.linspace(-3.0, 3.5, 481)
X, Z = np.meshgrid(x, z)
phi_sphere = sphere_potential_cartesian(X, Z, q=q, R=R, a=a)
phi_free = k_c * q / np.sqrt(X**2 + (Z - a)**2)

mask_inside = X**2 + Z**2 < R**2
mask_real = np.sqrt(X**2 + (Z - a)**2) < 0.10
phi_sphere = np.ma.array(phi_sphere, mask=(mask_inside | mask_real))
phi_free = np.ma.array(phi_free, mask=(mask_inside | mask_real))

xq = np.linspace(-3.0, 3.0, 31)
zq = np.linspace(-3.0, 3.3, 33)
Xq, Zq = np.meshgrid(xq, zq)
Exq, Ezq = sphere_field_cartesian(Xq, Zq, q=q, R=R, a=a)
mask_quiver = (Xq**2 + Zq**2 < R**2) | (np.sqrt(Xq**2 + (Zq - a)**2) < 0.22)
Exq = np.ma.array(Exq, mask=mask_quiver)
Ezq = np.ma.array(Ezq, mask=mask_quiver)

theta = np.linspace(0.0, np.pi, 800)
sigma_vals = sphere_sigma(theta, q=q, R=R, a=a)
q_induced_numeric = np.trapz(2.0 * np.pi * R**2 * sigma_vals * np.sin(theta), theta)

boundary_theta = np.linspace(0.0, np.pi, 600)
boundary_x = R * np.sin(boundary_theta)
boundary_z = R * np.cos(boundary_theta)
boundary_phi = sphere_potential_cartesian(boundary_x, boundary_z, q=q, R=R, a=a)

z_axis = np.linspace(R + 0.02, a - 0.12, 400)
phi_axis_sphere = sphere_potential_cartesian(np.zeros_like(z_axis), z_axis, q=q, R=R, a=a)
phi_axis_free = k_c * q / np.abs(z_axis - a)
force_sphere = q * q_image / (a - b) ** 2

fig, axes = plt.subplots(2, 2, figsize=(13, 10))

levels = np.linspace(-0.25, 1.5, 18)
axes[0, 0].contour(X, Z, phi_sphere, levels=levels, cmap="viridis")
axes[0, 0].add_patch(Circle((0, 0), R, fill=False, color="black", linewidth=2))
axes[0, 0].scatter([0.0], [a], color="crimson", s=70, label="real charge q")
axes[0, 0].scatter([0.0], [b], color="navy", s=70, label="image charge q'")
axes[0, 0].set_title("Potential outside the grounded sphere")
axes[0, 0].set_xlabel("x / R")
axes[0, 0].set_ylabel("z / R")
axes[0, 0].set_xlim(-3, 3)
axes[0, 0].set_ylim(-3, 3.5)
axes[0, 0].set_aspect("equal", adjustable="box")
axes[0, 0].legend(loc="upper left")

axes[0, 1].contour(X, Z, phi_sphere, levels=np.linspace(0.0, 1.5, 8), colors="0.75", linewidths=0.8)
axes[0, 1].quiver(Xq, Zq, Exq, Ezq, pivot="mid", color="tab:blue", scale=28)
axes[0, 1].add_patch(Circle((0, 0), R, fill=False, color="black", linewidth=2))
axes[0, 1].scatter([0.0], [a], color="crimson", s=70)
axes[0, 1].set_title("Electric field in the exterior region")
axes[0, 1].set_xlabel("x / R")
axes[0, 1].set_ylabel("z / R")
axes[0, 1].set_xlim(-3, 3)
axes[0, 1].set_ylim(-3, 3.5)
axes[0, 1].set_aspect("equal", adjustable="box")

axes[1, 0].plot(theta, sigma_vals, color="tab:green", linewidth=2)
axes[1, 0].axhline(0.0, color="black", linewidth=1)
axes[1, 0].set_title("Induced surface charge density on the sphere")
axes[1, 0].set_xlabel(r"$\theta$")
axes[1, 0].set_ylabel(r"$\sigma(\theta)$ in normalized units")
axes[1, 0].set_xlim(0, np.pi)

axes[1, 1].plot(z_axis, phi_axis_free, color="0.5", linestyle="--", linewidth=2, label="without conductor")
axes[1, 1].plot(z_axis, phi_axis_sphere, color="tab:purple", linewidth=2, label="with grounded sphere")
axes[1, 1].axvline(R, color="black", linewidth=1)
axes[1, 1].set_title("Potential along the symmetry axis")
axes[1, 1].set_xlabel("z / R")
axes[1, 1].set_ylabel(r"$\phi(0,z)$")
axes[1, 1].legend(loc="upper left")

plt.tight_layout()
plt.show()

print(f"Image charge q' = {q_image:.6f}")
print(f"Image position b = {b:.6f}")
print(f"Maximum |phi| on the spherical boundary: {np.max(np.abs(boundary_phi)):.6e}")
print(f"Numerical induced charge on the sphere: {q_induced_numeric:.6f}")
print(f"Exact induced charge: {q_image:.6f}")
print(f"Normalized force on the external charge: {force_sphere:.6f} z-hat")
```

The numerical boundary check confirms that the spherical surface is an equipotential to plotting accuracy. The induced charge density is sharply peaked near $\theta=0$, where the external charge is closest to the conductor, and becomes weaker on the far side. The comparison between the free-space potential and the grounded-sphere potential along the symmetry axis shows how the conductor suppresses the potential so as to enforce $\phi=0$ at $r=R$. The image charge is smaller in magnitude than the real charge because the grounded sphere need not acquire total charge $-q$; instead, its total induced charge is $-qR/a$.

## Advanced Conceptual Remarks

The image charges are not physically present. They generally lie inside the conductor or in some other excluded auxiliary region, and the electrostatic field they generate there need not have any direct physical meaning for the original problem. Only the potential and field in the physical region are relevant. This distinction between the physical region and the auxiliary region is indispensable; confusion on this point is one of the most common misunderstandings surrounding the method.

The method works only in specially symmetric geometries because one must solve a stringent functional problem: the candidate potential must satisfy an entire boundary condition, not merely match it at a few points. For generic conductor shapes there is no finite set of image charges that accomplishes this. One then turns to separation of variables, Green's functions in more general form, integral equation methods, or numerical solvers.

Boundary conditions select the image construction. Changing from grounded to isolated conductors, or from Dirichlet to Neumann data, changes the admissible Green's function and therefore the auxiliary sources required. Even in symmetric settings, the correct image system depends delicately on what is being held fixed physically.

The method of images is therefore both powerful and limited. Its power lies in turning difficult boundary value problems into exact algebraic constructions when symmetry permits. Its limitation lies in the rarity of geometries for which this can be done explicitly.

## Key Takeaways

- The method of images is justified by the uniqueness theorem, not by any literal replacement of a conductor by real charges.
- Image charges must lie outside the physical region so that the candidate potential has the correct source structure where the physical solution is defined.
- For the grounded plane, the image is a reflected charge of equal magnitude and opposite sign.
- For the grounded sphere, the image location and magnitude are determined by spherical inversion: $q'=-qR/a$ and $b=R^2/a$.
- The induced surface charge density and the force on the real charge follow directly once the potential is known.
- The method can be reinterpreted as the construction of a Dirichlet Green's function adapted to the boundary.

## Common Pitfalls

- Treating image charges as physical charges rather than auxiliary mathematical devices.
- Forgetting that the constructed potential needs to be valid only in the physical region.
- Using the image force formula inside the conductor or inferring unphysical fields in the auxiliary region.
- Neglecting the asymptotic condition at infinity in unbounded problems.
- Assuming that every conductor geometry admits a simple image solution.

## Connection to Later Topics

The method of images anticipates several later themes in graduate electrodynamics: Green's functions, boundary value problems for elliptic equations, uniqueness theorems, multipole expansions of induced charge distributions, conductor polarization, and more general analytical and numerical methods for solving Poisson's equation. It is therefore best understood not as an isolated trick, but as an exact and highly structured instance of electrostatic boundary value theory.
