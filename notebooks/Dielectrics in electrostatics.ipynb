# Dielectrics in Electrostatics

Dielectric media are central to electrodynamics because they provide the simplest nontrivial example of how matter responds to electromagnetic fields. Even in the static limit, the presence of polarizable material modifies charge distributions, boundary conditions, energy storage, and the relation between source charges and fields. The theory of dielectrics therefore occupies a foundational position between vacuum electrostatics and the more general electrodynamics of continuous media.

## Motivation and Physical Picture

A dielectric is a material that contains charges bound to atoms, molecules, or more complicated microscopic structures, but lacks the freely mobile carriers that characterize a conductor. In electrostatics a conductor rearranges charge until the electric field in its interior vanishes, whereas a dielectric generally supports a nonzero internal field and responds instead by a displacement of positive and negative charge centers relative to one another.

This displacement produces polarization. Microscopically one may imagine either the distortion of an already existing charge cloud or the reorientation of permanent molecular dipoles. Macroscopically the result is that the medium develops an electric dipole moment density. The induced charge associated with this rearrangement is called **bound charge**, in contrast with **free charge**, which denotes externally supplied or otherwise mobile charge that appears explicitly in macroscopic source descriptions.

The distinction is not merely terminological. If one writes Gauss's law directly in terms of the total charge density,

$$
\nabla\cdot \mathbf{E}=\frac{\rho_{\mathrm{tot}}}{\varepsilon_0}
=\frac{\rho_f+\rho_b}{\varepsilon_0},
$$

then the bound contribution is entangled with the medium response. It is therefore natural to introduce new field variables, especially the polarization $\mathbf{P}$ and displacement $\mathbf{D}$, that encode the material response in a more systematic way.

## Polarization Field

The macroscopic polarization field $\mathbf{P}(\mathbf{r})$ is defined as electric dipole moment per unit volume. If a coarse-grained volume element $\Delta V$ contains many microscopic dipoles whose total moment is $\Delta \mathbf{p}$, then

$$
\mathbf{P}=\lim_{\Delta V\to \text{macroscopic small}}\frac{\Delta \mathbf{p}}{\Delta V}.
$$

Its physical meaning is therefore direct: $\mathbf{P}$ points in the direction of positive charge displacement relative to negative charge displacement, and its magnitude measures the dipole moment density carried by the material.

To derive the effective charge density associated with a polarized medium, begin with the potential of a localized dipole distribution:

$$
\phi(\mathbf{r})=\frac{1}{4\pi\varepsilon_0}\int \mathbf{P}(\mathbf{r}')\cdot \nabla'\frac{1}{|\mathbf{r}-\mathbf{r}'|}\,d^3r'.
$$

Using

$$
\mathbf{P}\cdot \nabla'\frac{1}{R}
=\nabla'\cdot\left(\frac{\mathbf{P}}{R}\right)-\frac{\nabla'\cdot\mathbf{P}}{R},
\qquad R=|\mathbf{r}-\mathbf{r}'|,
$$

and applying the divergence theorem gives

$$
\phi(\mathbf{r})=\frac{1}{4\pi\varepsilon_0}
\left[
\oint \frac{\mathbf{P}\cdot \hat{\mathbf{n}}'}{|\mathbf{r}-\mathbf{r}'|}\,da'
-\int \frac{\nabla'\cdot \mathbf{P}}{|\mathbf{r}-\mathbf{r}'|}\,d^3r'
\right].
$$

Comparison with the Coulomb potential shows that the polarized medium behaves as though it carried the effective bound charge densities

$$
\rho_b=-\nabla\cdot \mathbf{P},
$$

and

$$
\sigma_b=\mathbf{P}\cdot \hat{\mathbf{n}}.
$$

These formulas admit a simple local interpretation. Consider a small pillbox-shaped volume element in a uniformly polarized material. Dipoles crossing the side walls cancel in pairs, while the faces normal to $\mathbf{P}$ acquire uncompensated charge. Spatial variation of $\mathbf{P}$ then produces incomplete cancellation from one volume element to the next, yielding the bound volume charge density $-\nabla\cdot\mathbf{P}$. The surface formula is the boundary analogue of the same counting argument.

## Electric Displacement Field

Maxwell's electrostatic Gauss law for the total charge is

$$
\nabla\cdot\mathbf{E}=\frac{\rho_f+\rho_b}{\varepsilon_0}.
$$

Substituting $\rho_b=-\nabla\cdot\mathbf{P}$ gives

$$
\nabla\cdot\left(\varepsilon_0\mathbf{E}+\mathbf{P}\right)=\rho_f.
$$

This motivates the definition of the displacement field

$$
\mathbf{D}=\varepsilon_0\mathbf{E}+\mathbf{P},
$$

for which Gauss's law in matter becomes

$$
\nabla\cdot\mathbf{D}=\rho_f.
$$

The utility of $\mathbf{D}$ is therefore organizational: it separates externally specified or mobile charge from the induced polarization charge hidden in the medium response. In many boundary value problems the free charge is simpler to describe than the total charge, and $\mathbf{D}$ packages the complicated bound contribution into the constitutive properties of the material.

At the same time one must not overinterpret $\mathbf{D}$. Unlike $\mathbf{E}$, it is not a microscopic force field. It is a derived macroscopic quantity whose usefulness depends on the constitutive description of matter. In anisotropic, nonlinear, dispersive, or spatially nonlocal media, the relation between $\mathbf{D}$ and $\mathbf{E}$ may become substantially more complicated than the elementary form used in simple dielectric theory.

## Linear Dielectrics

For a linear isotropic dielectric one assumes the constitutive relation

$$
\mathbf{P}=\varepsilon_0\chi_e \mathbf{E},
$$

where $\chi_e$ is the electric susceptibility. Then

$$
\mathbf{D}=\varepsilon_0\mathbf{E}+\mathbf{P}
=\varepsilon_0(1+\chi_e)\mathbf{E}
=\varepsilon \mathbf{E},
$$

with

$$
\varepsilon=\varepsilon_0(1+\chi_e)=\varepsilon_0\varepsilon_r.
$$

Here $\varepsilon_r$ is the relative permittivity. The assumptions behind this form are substantial:

- **Linear** means the response is proportional to the applied field.
- **Isotropic** means no direction in the material is distinguished.
- **Homogeneous** means the constitutive parameters are independent of position.

Many real materials violate one or more of these assumptions. Crystals are often anisotropic, strong fields can produce nonlinear polarization, and composite materials may be strongly inhomogeneous. Nevertheless, the linear isotropic homogeneous model is the essential first approximation and underlies a large fraction of standard electrostatic applications.

## Poisson's Equation in Dielectrics

In electrostatics one may still write

$$
\mathbf{E}=-\nabla\phi.
$$

For general dielectric media the equation $\nabla\cdot\mathbf{D}=\rho_f$ becomes

$$
\nabla\cdot\left(\varepsilon_0\nabla(-\phi)+\mathbf{P}\right)=\rho_f.
$$

If a constitutive relation $\mathbf{D}=\varepsilon\mathbf{E}$ is available, then

$$
\nabla\cdot(\varepsilon \nabla \phi)=-\rho_f.
$$

This is the generalized Poisson equation for an inhomogeneous dielectric.

When $\varepsilon$ is uniform, it reduces to

$$
\varepsilon \nabla^2\phi=-\rho_f,
\qquad\text{or}\qquad
\nabla^2\phi=-\frac{\rho_f}{\varepsilon}.
$$

The physical interpretation is clear: the dielectric weakens the electric field generated by a given free-charge distribution relative to vacuum, because part of the total field is screened by polarization. In the inhomogeneous case the operator itself changes, and the spatial variation of $\varepsilon$ influences the field even where the free charge density vanishes. This is why dielectric interfaces require separate boundary conditions rather than a simple global rescaling of Coulomb's law.

## Boundary Conditions at Dielectric Interfaces

Let an interface separate media 1 and 2, and let $\hat{\mathbf{n}}$ point from medium 1 into medium 2.

Applying Gauss's law for $\mathbf{D}$ to a thin pillbox straddling the interface gives

$$
\hat{\mathbf{n}}\cdot(\mathbf{D}_2-\mathbf{D}_1)=\sigma_f,
$$

where $\sigma_f$ is the free surface charge density. Thus the normal component of $\mathbf{D}$ is discontinuous only if free surface charge is present.

Applying Faraday's law in electrostatics,

$$
\nabla\times \mathbf{E}=0,
$$

to a narrow rectangular loop crossing the interface yields

$$
\hat{\mathbf{n}}\times(\mathbf{E}_2-\mathbf{E}_1)=0.
$$

Hence the tangential component of $\mathbf{E}$ is continuous across the interface.

For linear isotropic media this implies that the normal component of $\mathbf{E}$ is generally discontinuous when $\varepsilon$ changes, while the tangential component of $\mathbf{D}$ is generally discontinuous. These statements are simply different projections of the same boundary conditions once $\mathbf{D}=\varepsilon\mathbf{E}$ is imposed. Physically, dielectric interfaces support polarization charge that alters the normal field while leaving the electrostatic tangential continuity intact.

## Energy in Dielectrics

The electrostatic energy density in matter is

$$
u=\frac{1}{2}\mathbf{E}\cdot\mathbf{D}.
$$

This reduces to $\frac{1}{2}\varepsilon_0 E^2$ in vacuum and to $\frac{1}{2}\varepsilon E^2$ in a linear isotropic dielectric. The presence of a dielectric therefore changes the amount of energy stored for a given electric field configuration.

For capacitor problems the consequence is immediate. If a dielectric fills the space between plates while the free charge is held fixed, the electric field decreases and so does the potential difference, changing the capacitance and the stored energy. If instead the voltage is held fixed by a battery, the free charge adjusts. Energy methods must therefore be handled with care: the correct thermodynamic interpretation depends on what external constraints are maintained.

## Canonical Examples

### 1. Uniformly polarized slab

Take a slab with uniform polarization $\mathbf{P}=P_0\hat{\mathbf{z}}$. Then $\nabla\cdot\mathbf{P}=0$ in the bulk, so there is no bound volume charge. However, the faces normal to $\hat{\mathbf{z}}$ carry

$$
\sigma_b=\pm P_0.
$$

The slab therefore behaves like two oppositely charged sheets.

### 2. Uniformly polarized sphere

For a sphere with constant polarization $\mathbf{P}=P_0\hat{\mathbf{z}}$, again $\rho_b=0$ in the interior, but the surface carries

$$
\sigma_b(\theta)=P_0\cos\theta.
$$

This produces a uniform internal field and an external dipole field.

### 3. Dielectric filling a parallel-plate capacitor

If a linear dielectric fills the gap between ideal large parallel plates carrying free surface charge densities $\pm \sigma_f$, then

$$
\mathbf{D}=\sigma_f \hat{\mathbf{n}},
\qquad
\mathbf{E}=\frac{\sigma_f}{\varepsilon}\hat{\mathbf{n}}.
$$

Compared with vacuum, the field is reduced by the factor $\varepsilon_r$.

### 4. Interface between two linear dielectrics

With no free surface charge,

$$
\hat{\mathbf{n}}\cdot(\mathbf{D}_2-\mathbf{D}_1)=0
\quad\Rightarrow\quad
\varepsilon_1 E_{1n}=\varepsilon_2 E_{2n}.
$$

The normal electric field therefore changes inversely with permittivity.

### 5. Point charge in a uniform dielectric medium

For uniform $\varepsilon$, Coulomb's law becomes

$$
\phi(\mathbf{r})=\frac{1}{4\pi\varepsilon}\frac{q}{r},
\qquad
\mathbf{E}=\frac{1}{4\pi\varepsilon}\frac{q}{r^2}\hat{\mathbf{r}}.
$$

This is not a new law of electrostatics but the vacuum law combined with a simple constitutive response.

In all these examples the central theme is the same: the dielectric alters fields by generating bound charge through polarization.

## Polarized Sphere

The uniformly polarized sphere is one of the most important exactly solvable problems in electrostatics with matter. Let a sphere of radius $R$ have constant polarization

$$
\mathbf{P}=P_0\hat{\mathbf{z}}.
$$

Since $\mathbf{P}$ is uniform,

$$
\rho_b=-\nabla\cdot\mathbf{P}=0
$$

throughout the interior, while on the surface

$$
\sigma_b=\mathbf{P}\cdot \hat{\mathbf{n}}=P_0\cos\theta.
$$

The potential may be obtained by solving Laplace's equation inside and outside with this surface charge. The result is

$$
\phi_{\mathrm{in}}(r,\theta)=\frac{P_0}{3\varepsilon_0}r\cos\theta,
$$

and

$$
\phi_{\mathrm{out}}(r,\theta)=\frac{1}{4\pi\varepsilon_0}\frac{\mathbf{p}\cdot\hat{\mathbf{r}}}{r^2},
\qquad
\mathbf{p}=\frac{4\pi R^3}{3}\mathbf{P}.
$$

Hence the external field is exactly that of a point dipole at the center with dipole moment equal to the total dipole moment of the sphere. The interior field is uniform:

$$
\mathbf{E}_{\mathrm{in}}=-\nabla\phi_{\mathrm{in}}
=-\frac{P_0}{3\varepsilon_0}\hat{\mathbf{z}}.
$$

This example is important because it links bound surface charge, multipole structure, and the macroscopic meaning of polarization in a clean and analytically exact way.

## Dielectric Sphere in an External Field

Now consider a dielectric sphere of radius $R$ and permittivity $\varepsilon$ placed in a uniform external field

$$
\mathbf{E}_0=E_0\hat{\mathbf{z}}.
$$

The potential satisfies Laplace's equation both inside and outside. Axial symmetry implies that only the $\ell=1$ Legendre sector is needed:

$$
\phi_{\mathrm{in}}=-Ar\cos\theta,
$$

and

$$
\phi_{\mathrm{out}}=-E_0 r\cos\theta+\frac{B\cos\theta}{r^2}.
$$

Applying continuity of $\phi$ at $r=R$ and continuity of the normal component of $\mathbf{D}$ in the absence of free surface charge,

$$
\varepsilon \left.\frac{\partial \phi_{\mathrm{in}}}{\partial r}\right|_{R}
=\varepsilon_0 \left.\frac{\partial \phi_{\mathrm{out}}}{\partial r}\right|_{R},
$$

one finds

$$
A=\frac{3\varepsilon_0}{\varepsilon+2\varepsilon_0}E_0,
$$

and

$$
B=E_0 R^3\frac{\varepsilon-\varepsilon_0}{\varepsilon+2\varepsilon_0}.
$$

Therefore

$$
\phi_{\mathrm{in}}=
-\frac{3\varepsilon_0}{\varepsilon+2\varepsilon_0}E_0 r\cos\theta,
$$

and

$$
\phi_{\mathrm{out}}=
-E_0 r\cos\theta
+E_0 R^3\frac{\varepsilon-\varepsilon_0}{\varepsilon+2\varepsilon_0}\frac{\cos\theta}{r^2}.
$$

The second term outside is a dipole potential, so the induced dipole moment is

$$
\mathbf{p}_{\mathrm{ind}}
=4\pi\varepsilon_0 R^3
\frac{\varepsilon-\varepsilon_0}{\varepsilon+2\varepsilon_0}\mathbf{E}_0.
$$

For weak susceptibility, $\varepsilon=\varepsilon_0(1+\chi_e)$ with $|\chi_e|\ll 1$, the dipole moment is proportional to $\chi_e$. In the conductor limit $\varepsilon/\varepsilon_0\to\infty$, one obtains

$$
\mathbf{p}_{\mathrm{ind}}\to 4\pi\varepsilon_0 R^3 \mathbf{E}_0,
$$

and the internal field tends to zero, as expected for a conducting sphere in electrostatic equilibrium.

## Microscopic to Macroscopic View

The macroscopic polarization field is a coarse-grained description. At the microscopic level matter contains discrete nuclei and electrons; the microscopic electric field fluctuates violently over atomic distances. Macroscopic electrodynamics smooths these variations by averaging over volumes large compared with microscopic scales but small compared with the spatial scale on which the external fields vary appreciably.

This averaging justifies the use of continuous fields such as $\mathbf{P}$, $\mathbf{E}$, and $\mathbf{D}$. The relation between $\mathbf{P}$ and $\mathbf{E}$ can be motivated from molecular polarizability: an individual molecule may acquire an induced dipole moment $\mathbf{p}_{\mathrm{mol}}=\alpha \mathbf{E}_{\mathrm{loc}}$. After averaging, one obtains a polarization density. However, the local field acting on a molecule is not generally identical to the macroscopic field, which is why microscopic derivations of constitutive relations require care. For present purposes the important point is that macroscopic dielectric theory is an averaged description, not a literal statement about pointwise microscopic charge motion.

## Computational Demonstration: Bound Charge on a Polarized Sphere

For a uniformly polarized sphere with $\mathbf{P}=P_0\hat{\mathbf{z}}$, the surface bound charge density is $\sigma_b(\theta)=P_0\cos\theta$. The following code plots this distribution on the sphere and as a one-dimensional function of polar angle. It also compares the exact internal field with the theoretical value $-P_0/(3\varepsilon_0)\,\hat{\mathbf{z}}$ in normalized units.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

plt.rcParams.update({
    "figure.figsize": (12, 9),
    "font.size": 11,
    "axes.grid": False,
})

eps0 = 1.0  # normalized permittivity for visualization

P0 = 1.0
R = 1.0

theta = np.linspace(0.0, np.pi, 600)
sigma_b = P0 * np.cos(theta)

# Parametric boundary of the sphere in the x-z plane.
x_boundary = R * np.sin(theta)
z_boundary = R * np.cos(theta)

# Interior field predicted by the uniformly polarized sphere solution.
E_inside_z = -P0 / (3.0 * eps0)

fig, axes = plt.subplots(1, 2, figsize=(13, 5.5))

scatter = axes[0].scatter(
    x_boundary,
    z_boundary,
    c=sigma_b,
    cmap="coolwarm",
    s=20,
    vmin=-P0,
    vmax=P0,
)
axes[0].add_patch(Circle((0, 0), R, fill=False, color="black", linewidth=1.5))
axes[0].arrow(
    0.0, 0.0, 0.0, 0.7, width=0.02, head_width=0.08, head_length=0.1, color="black"
)
axes[0].text(0.08, 0.72, "P", fontsize=12)
axes[0].set_title("Bound surface charge on a uniformly polarized sphere")
axes[0].set_xlabel("x / R")
axes[0].set_ylabel("z / R")
axes[0].set_aspect("equal", adjustable="box")
axes[0].set_xlim(-1.2, 1.2)
axes[0].set_ylim(-1.2, 1.2)
plt.colorbar(scatter, ax=axes[0], label=r"$\sigma_b / P_0$")

axes[1].plot(theta, sigma_b, color="tab:red", linewidth=2)
axes[1].axhline(0.0, color="black", linewidth=1)
axes[1].set_title("Surface charge density versus polar angle")
axes[1].set_xlabel(r"$\theta$")
axes[1].set_ylabel(r"$\sigma_b(\theta)$")
axes[1].set_xlim(0, np.pi)

plt.tight_layout()
plt.show()

print(f"Uniform internal field: E_z = {E_inside_z:.6f} in normalized units")
print(f"North-pole surface charge density: {sigma_b[0]:.6f}")
print(f"Equatorial surface charge density: {sigma_b[len(theta)//2]:.6f}")
print(f"South-pole surface charge density: {sigma_b[-1]:.6f}")
```

The plot makes the angular structure of the bound surface charge explicit. Positive charge accumulates on the hemisphere toward which $\mathbf{P}$ points, negative charge on the opposite hemisphere, and the equator carries zero bound surface charge. The example is a concrete illustration that a uniform polarization field need not generate any bound volume charge, yet can still produce a nontrivial electric field through its surface charge.

## Computational Demonstration: Dielectric Sphere in an External Field

The dielectric sphere in a uniform external field is a classical boundary value problem. The code below evaluates the exact potential inside and outside the sphere, visualizes equipotentials and field lines in a meridional cross section, and extracts the induced dipole moment. We use normalized units with $\varepsilon_0=1$ and choose a moderate dielectric contrast so that the interior field reduction is visually clear.

```python
def dielectric_sphere_coefficients(eps_rel=4.0, E0=1.0, R=1.0):
    eps_out = 1.0
    eps_in = eps_rel * eps_out
    A = 3.0 * eps_out / (eps_in + 2.0 * eps_out) * E0
    B = E0 * R**3 * (eps_in - eps_out) / (eps_in + 2.0 * eps_out)
    return eps_in, eps_out, A, B


def sphere_external_field_potential(X, Z, eps_rel=4.0, E0=1.0, R=1.0):
    eps_in, eps_out, A, B = dielectric_sphere_coefficients(eps_rel=eps_rel, E0=E0, R=R)
    r = np.sqrt(X**2 + Z**2)
    cos_theta = np.divide(Z, r, out=np.zeros_like(r), where=r > 0)
    phi_in = -A * r * cos_theta
    phi_out = -E0 * r * cos_theta + B * cos_theta / np.where(r > 0, r**2, np.inf)
    return np.where(r <= R, phi_in, phi_out), A, B


eps_rel = 4.0
E0 = 1.0
R = 1.0
eps_in, eps_out, A, B = dielectric_sphere_coefficients(eps_rel=eps_rel, E0=E0, R=R)
p_ind = 4.0 * np.pi * eps_out * B

x = np.linspace(-3.0, 3.0, 401)
z = np.linspace(-3.0, 3.0, 401)
X, Z = np.meshgrid(x, z)
phi, A, B = sphere_external_field_potential(X, Z, eps_rel=eps_rel, E0=E0, R=R)

# Numerical field from the potential for visualization.
dz = z[1] - z[0]
dx = x[1] - x[0]
Ez, Ex = np.gradient(-phi, dz, dx)

fig, axes = plt.subplots(1, 2, figsize=(13, 5.5))

levels = np.linspace(-3.0, 3.0, 19)
cont = axes[0].contour(X, Z, phi, levels=levels, cmap="coolwarm")
axes[0].clabel(cont, inline=True, fontsize=8, fmt="%.1f")
axes[0].add_patch(Circle((0, 0), R, fill=False, color="black", linewidth=2))
axes[0].set_title("Potential of a dielectric sphere in a uniform field")
axes[0].set_xlabel("x / R")
axes[0].set_ylabel("z / R")
axes[0].set_aspect("equal", adjustable="box")

skip = 14
axes[1].streamplot(
    x, z, Ex, Ez, color=np.hypot(Ex, Ez), density=1.2, cmap="viridis", linewidth=1
)
axes[1].add_patch(Circle((0, 0), R, fill=False, color="black", linewidth=2))
axes[1].set_title("Electric field in the presence of the dielectric sphere")
axes[1].set_xlabel("x / R")
axes[1].set_ylabel("z / R")
axes[1].set_aspect("equal", adjustable="box")

plt.tight_layout()
plt.show()

print(f"Interior field magnitude along +z: {A:.6f}")
print(f"Induced dipole moment magnitude: {p_ind:.6f}")
print(f"Conductor-limit dipole moment for the same R and E0: {4*np.pi*R**3*E0:.6f}")
```

The interior equipotentials are nearly parallel, reflecting the fact that the exact internal field is uniform. Outside the sphere, the field is the superposition of the applied uniform field and the induced dipole field. As the dielectric contrast increases, the field lines bulge more strongly around the sphere and the induced dipole moment approaches the conductor limit.

## Computational Demonstration: Piecewise-Constant Permittivity in One Dimension

The generalized Poisson equation

$$
\frac{d}{dx}\left(\varepsilon(x)\frac{d\phi}{dx}\right)=-\rho_f(x)
$$

is especially transparent in one dimension. The code below solves a simple finite-difference problem with piecewise constant permittivity and localized free charge. It then plots the potential, electric field, and displacement field, allowing one to see directly how the dielectric interface alters the slope of the potential while preserving the expected interface conditions.

```python
# One-dimensional finite-difference solution of d/dx(epsilon dphi/dx) = -rho_f.
L = 1.0
N = 500
x = np.linspace(0.0, L, N)
dx = x[1] - x[0]

# Piecewise-constant permittivity.
eps_left = 1.0
eps_right = 4.0
epsilon = np.where(x < 0.5, eps_left, eps_right)

# Localized free charge represented by a narrow Gaussian.
rho_f = 8.0 * np.exp(-((x - 0.25) / 0.04) ** 2)

# Dirichlet boundary conditions.
phi_left = 0.0
phi_right = 0.0

# Interface permittivities at half-grid points.
eps_half = 0.5 * (epsilon[:-1] + epsilon[1:])

A_mat = np.zeros((N, N))
b_vec = np.zeros(N)

A_mat[0, 0] = 1.0
b_vec[0] = phi_left
A_mat[-1, -1] = 1.0
b_vec[-1] = phi_right

for i in range(1, N - 1):
    eps_m = eps_half[i - 1]
    eps_p = eps_half[i]
    A_mat[i, i - 1] = eps_m / dx**2
    A_mat[i, i] = -(eps_m + eps_p) / dx**2
    A_mat[i, i + 1] = eps_p / dx**2
    b_vec[i] = -rho_f[i]

phi_1d = np.linalg.solve(A_mat, b_vec)
E_1d = -np.gradient(phi_1d, dx)
D_1d = epsilon * E_1d

fig, axes = plt.subplots(3, 1, figsize=(11, 9), sharex=True)

axes[0].plot(x, phi_1d, color="tab:blue", linewidth=2)
axes[0].axvline(0.5, color="black", linestyle="--", linewidth=1)
axes[0].set_ylabel(r"$\phi(x)$")
axes[0].set_title("1D dielectric Poisson problem with piecewise permittivity")

axes[1].plot(x, E_1d, color="tab:orange", linewidth=2, label=r"$E_x$")
axes[1].plot(x, D_1d, color="tab:green", linewidth=2, label=r"$D_x$")
axes[1].axvline(0.5, color="black", linestyle="--", linewidth=1)
axes[1].legend(loc="upper right")
axes[1].set_ylabel("Field amplitude")

axes[2].plot(x, epsilon, color="tab:red", linewidth=2)
axes[2].axvline(0.5, color="black", linestyle="--", linewidth=1)
axes[2].set_ylabel(r"$\varepsilon(x)$")
axes[2].set_xlabel("x")

plt.tight_layout()
plt.show()

interface_index = np.argmin(np.abs(x - 0.5))
print(f"Potential near interface from left/right: {phi_1d[interface_index-1]:.6f}, {phi_1d[interface_index+1]:.6f}")
print(f"E near interface from left/right: {E_1d[interface_index-1]:.6f}, {E_1d[interface_index+1]:.6f}")
print(f"D near interface from left/right: {D_1d[interface_index-1]:.6f}, {D_1d[interface_index+1]:.6f}")
```

The potential remains continuous, as expected in electrostatics. The electric field generally changes across the interface because the permittivity changes, while the displacement field is much closer to continuity away from the localized free-charge region. This is the one-dimensional manifestation of the boundary condition $\hat{\mathbf{n}}\cdot(\mathbf{D}_2-\mathbf{D}_1)=\sigma_f$.

## Advanced Conceptual Remarks

Bound charge is not an independent new species of charge. It is ordinary electric charge appearing in a reorganized form once the microscopic constituents of matter respond to an applied field. The split

$$
\rho_{\mathrm{tot}}=\rho_f+\rho_b
$$

is therefore a bookkeeping distinction within macroscopic theory, not a statement that nature contains two kinds of electric charge.

The constitutive relation $\mathbf{P}\propto \mathbf{E}$ breaks down when the response becomes nonlinear, anisotropic, hysteretic, frequency dependent, or nonlocal. Ferroelectrics, strongly inhomogeneous composites, and many optical materials lie beyond the scope of the elementary linear dielectric model.

Nevertheless, the simple theory developed here is foundational. It provides the conceptual bridge from vacuum electrostatics to electromagnetic waves in matter, refractive index, polarization currents, and the broader continuum theory of media.

## Key Takeaways

- Polarization $\mathbf{P}$ is electric dipole moment per unit volume.
- Bound charge arises from polarization according to $\rho_b=-\nabla\cdot\mathbf{P}$ and $\sigma_b=\mathbf{P}\cdot\hat{\mathbf{n}}$.
- The displacement field $\mathbf{D}=\varepsilon_0\mathbf{E}+\mathbf{P}$ satisfies $\nabla\cdot\mathbf{D}=\rho_f$.
- In a linear isotropic dielectric, $\mathbf{D}=\varepsilon\mathbf{E}$ and the field equations are modified accordingly.
- Dielectric interfaces are governed by continuity of tangential $\mathbf{E}$ and the free-charge jump condition for normal $\mathbf{D}$.
- Dielectrics alter electrostatic energy storage and naturally enter capacitor and boundary value problems.

## Common Pitfalls

- Confusing free charge with total charge or treating bound charge as a separate fundamental charge species.
- Assuming $\mathbf{D}$ is more physically fundamental than $\mathbf{E}$ rather than a constitutive convenience.
- Forgetting that a uniform polarization can produce zero bound volume charge but nonzero bound surface charge.
- Applying $\mathbf{P}=\varepsilon_0\chi_e\mathbf{E}$ outside its regime of validity.
- Ignoring interface conditions when the permittivity changes spatially.

## Connection to Later Topics

The theory of dielectrics leads directly to boundary value problems in matter, capacitance and energy methods, forces in polarizable media, multipole descriptions of induced charge, and eventually electromagnetic waves in continuous media. It also provides the static foundation for later discussions of dispersion, susceptibility tensors, and more general constitutive response.
