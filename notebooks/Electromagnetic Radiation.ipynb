# Electromagnetic Radiation

Electromagnetic radiation is one of the decisive themes of classical electrodynamics because it marks the transition from static field theory to genuinely dynamical propagation. In electrostatics and magnetostatics, the fields are tied instantaneously to time-independent sources or steady currents. By contrast, radiation arises from time-dependent charge and current distributions, propagates causally at finite speed, and transports energy and momentum away from the source region. The theory therefore combines Maxwell's equations, wave propagation, retarded solutions, and asymptotic analysis into a single framework.

## Motivation and Physical Context

A static charge distribution produces an electrostatic field determined by Coulomb's law, and a steady current distribution produces a magnetostatic field determined by the Biot-Savart law or Ampere's law. Neither case involves the continual transport of energy to infinity. Radiation appears only when the sources vary in time in such a way that the electromagnetic field develops a propagating component.

The physical statement that an accelerated charge radiates is more precise than a slogan. What matters is that the field contains an asymptotic contribution which remains finite when multiplied by the distance from the source and which therefore gives a nonvanishing outward energy flux through spheres of arbitrarily large radius. If the Poynting flux through a sphere of radius $r$ scales as $r^{-2}$, then the total power crossing the sphere remains finite as $r \to \infty$; this is the characteristic signature of radiation.

It is useful to distinguish three spatial regimes for localized, time-dependent sources:

- **Near fields** are dominated by terms falling as $1/r^3$. They retain a quasi-static character and are strongly tied to the instantaneous source configuration.
- **Induction fields** fall as $1/r^2$. They represent intermediate behavior associated with time-varying source moments but do not by themselves carry finite power to infinity.
- **Radiation fields** fall as $1/r$. These are transverse fields whose Poynting flux decays as $1/r^2$, thereby transporting finite energy outward.

These distinctions make clear why the full time-dependent Maxwell equations are essential. Once the sources vary in time, one must abandon the static picture and replace it with a causal propagation theory in which field values at $(\mathbf r,t)$ depend on the sources at earlier times.

## Maxwell Equations and Wave Equations

In vacuum, Maxwell's equations in SI units are

$$
\nabla \cdot \mathbf E = \frac{\rho}{\varepsilon_0},
\qquad
\nabla \cdot \mathbf B = 0,
$$

$$
\nabla \times \mathbf E = -\frac{\partial \mathbf B}{\partial t},
\qquad
\nabla \times \mathbf B = \mu_0 \mathbf J + \mu_0 \varepsilon_0 \frac{\partial \mathbf E}{\partial t}.
$$

The homogeneous equations suggest the introduction of scalar and vector potentials,

$$
\mathbf B = \nabla \times \mathbf A,
\qquad
\mathbf E = -\nabla \phi - \frac{\partial \mathbf A}{\partial t}.
$$

The potentials are not unique. Under the gauge transformation

$$
\mathbf A \mapsto \mathbf A + \nabla \chi,
\qquad
\phi \mapsto \phi - \frac{\partial \chi}{\partial t},
$$

the physical fields remain unchanged. For radiation theory the most natural choice is the **Lorenz gauge**,

$$
\nabla \cdot \mathbf A + \frac{1}{c^2}\frac{\partial \phi}{\partial t} = 0,
\qquad c = \frac{1}{\sqrt{\mu_0 \varepsilon_0}}.
$$

Substituting the potentials into Maxwell's equations and imposing the Lorenz gauge yields decoupled inhomogeneous wave equations:

$$
\nabla^2 \phi - \frac{1}{c^2}\frac{\partial^2 \phi}{\partial t^2} = -\frac{\rho}{\varepsilon_0},
$$

$$
\nabla^2 \mathbf A - \frac{1}{c^2}\frac{\partial^2 \mathbf A}{\partial t^2} = -\mu_0 \mathbf J.
$$

The formal resemblance to the scalar wave equation is essential. The electromagnetic potentials are not determined by instantaneous source values but by solutions of hyperbolic equations, so propagation occurs at the finite speed $c$. This causal structure is the mathematical foundation of radiation theory.

## Retarded Potentials

The Green's function of the wave operator in three spatial dimensions selects the retarded solution consistent with causality. For localized sources one obtains

$$
\phi(\mathbf r,t)=\frac{1}{4\pi\varepsilon_0}\int \frac{\rho(\mathbf r',t_r)}{|\mathbf r-\mathbf r'|}\,d^3r',
$$

$$
\mathbf A(\mathbf r,t)=\frac{\mu_0}{4\pi}\int \frac{\mathbf J(\mathbf r',t_r)}{|\mathbf r-\mathbf r'|}\,d^3r',
$$

where the retarded time is

$$
t_r=t-\frac{|\mathbf r-\mathbf r'|}{c}.
$$

These formulas state that the field observed at $(\mathbf r,t)$ depends on the source at the earlier time required for a signal to propagate from $\mathbf r'$ to $\mathbf r$ at speed $c$. This is not a calculational refinement but a structural necessity: the theory is causal. The source cannot influence the field at spacelike separated events.

The appearance of $t_r$ also shows why time-dependent electrodynamics is intrinsically nonlocal in time. One cannot simply insert the instantaneous charge and current into static formulas. Radiation theory begins precisely where this static intuition fails.

## Fields from Retarded Potentials

The electric and magnetic fields follow from

$$
\mathbf E = -\nabla \phi - \frac{\partial \mathbf A}{\partial t},
\qquad
\mathbf B = \nabla \times \mathbf A.
$$

Because the sources appear at retarded time, differentiating the potentials requires care: both the explicit factor $1/|\mathbf r-\mathbf r'|$ and the implicit dependence through $t_r$ contribute. The exact results may be organized into Jefimenko-type expressions, for example

$$
\mathbf E(\mathbf r,t)=\frac{1}{4\pi\varepsilon_0}\int \left[\frac{\rho(\mathbf r',t_r)}{R^2}\hat{\mathbf R}+\frac{1}{cR}\frac{\partial \rho(\mathbf r',t_r)}{\partial t}\hat{\mathbf R}-\frac{1}{c^2R}\frac{\partial \mathbf J(\mathbf r',t_r)}{\partial t}\right]d^3r',
$$

$$
\mathbf B(\mathbf r,t)=\frac{\mu_0}{4\pi}\int \left[\frac{\mathbf J(\mathbf r',t_r)}{R^2}+\frac{1}{cR}\frac{\partial \mathbf J(\mathbf r',t_r)}{\partial t}\right]\times \hat{\mathbf R}\,d^3r',
$$

with $\mathbf R = \mathbf r-\mathbf r'$, $R=|\mathbf R|$, and $\hat{\mathbf R}=\mathbf R/R$.

These formulas exhibit the radial hierarchy directly. Terms proportional to $R^{-2}$ and stronger are associated with near and induction behavior, while the terms proportional to $R^{-1}$ involve time derivatives of the sources and become dominant in the far zone. The complexity of the exact fields is therefore not arbitrary; it reflects the coexistence of several physically distinct regimes within a single exact solution.

## Radiation Zone and Far-Field Approximation

Let the source be confined to a region of characteristic size $a$. The **radiation zone** or **far zone** is the regime in which

$$
r \gg a,
$$

and, for monochromatic sources of angular frequency $\omega$, usually also $r \gg \lambda = 2\pi c/\omega$ when one wants clean asymptotic separation of terms. In this regime the observation point is so far from the source that all source points are viewed under nearly the same direction.

One then expands

$$
R = |\mathbf r-\mathbf r'| \approx r-\hat{\mathbf r}\cdot \mathbf r',
\qquad
\frac{1}{R} \approx \frac{1}{r},
$$

while retaining the source-position dependence inside the retarded time,

$$
t_r \approx t-\frac{r}{c}+\frac{\hat{\mathbf r}\cdot \mathbf r'}{c}.
$$

This distinction is central. The amplitude factor may be approximated by $1/r$, but phase variations across the source can still matter. For sufficiently localized sources, the leading far-zone fields become transverse:

$$
\mathbf E_{\mathrm{rad}} \perp \hat{\mathbf r},
\qquad
\mathbf B_{\mathrm{rad}} = \frac{1}{c}\hat{\mathbf r}\times \mathbf E_{\mathrm{rad}}.
$$

The hierarchy of radial decay is then:

- near-field terms: $1/r^3$,
- induction terms: $1/r^2$,
- radiation terms: $1/r$.

Because the outward energy flux is quadratic in the fields, only the $1/r$ contribution yields a finite integrated power at large radius. This is the operational meaning of radiation.

## Dipole Radiation

For a localized source whose dimensions are small compared with the observation distance and, in the simplest long-wavelength limit, small compared with the wavelength, the leading radiative term is usually the electric dipole contribution. Define the electric dipole moment by

$$
\mathbf p(t)=\int \mathbf r'\,\rho(\mathbf r',t)\,d^3r'.
$$

In the far zone, the radiation field of a time-dependent dipole can be written in compact vector form as

$$
\mathbf E_{\mathrm{rad}}(\mathbf r,t)=\frac{1}{4\pi\varepsilon_0 c^2}\frac{\hat{\mathbf r}\times\left[\hat{\mathbf r}\times \ddot{\mathbf p}(t-r/c)\right]}{r},
$$

$$
\mathbf B_{\mathrm{rad}}(\mathbf r,t)=\frac{\mu_0}{4\pi}\frac{\hat{\mathbf r}\times \ddot{\mathbf p}(t-r/c)}{r} = \frac{1}{c}\hat{\mathbf r}\times \mathbf E_{\mathrm{rad}}.
$$

Hence radiation is controlled by the second time derivative of the dipole moment, not by the dipole moment itself. For a monochromatic dipole oriented along $\hat{\mathbf z}$,

$$
\mathbf p(t)=p_0\cos \omega t\,\hat{\mathbf z},
$$

the far-zone fields are purely transverse and have the angular form

$$
E_\theta(r,\theta,t)=\frac{\mu_0 \omega^2 p_0}{4\pi}\frac{\sin\theta}{r}\cos\!\left[\omega\left(t-\frac{r}{c}\right)\right],
$$

$$
B_\phi(r,\theta,t)=\frac{1}{c}E_\theta(r,\theta,t).
$$

The field vanishes on the dipole axis and is maximal in the equatorial plane. The radiation is linearly polarized, with the electric field in the local $\hat{\boldsymbol\theta}$ direction and the magnetic field in the $\hat{\boldsymbol\phi}$ direction. The transversality of the far field is exact at leading order in $1/r$.

## Radiated Power and the Poynting Vector

The electromagnetic energy flux is described by the Poynting vector,

$$
\mathbf S = \frac{1}{\mu_0}\mathbf E\times \mathbf B.
$$

In the far zone of dipole radiation, $\mathbf E$, $\mathbf B$, and $\hat{\mathbf r}$ form a mutually orthogonal triad, so the Poynting vector points radially outward:

$$
S_r = \frac{E_\theta^2}{\mu_0 c}.
$$

For the oscillating electric dipole $\mathbf p(t)=p_0\cos \omega t\,\hat{\mathbf z}$, the instantaneous angular distribution is

$$
\frac{dP}{d\Omega} = r^2 S_r
= \frac{\mu_0 \omega^4 p_0^2}{16\pi^2 c}\sin^2\theta\,\cos^2\!\left[\omega\left(t-\frac{r}{c}\right)\right].
$$

Time-averaging over a cycle gives

$$
\left\langle \frac{dP}{d\Omega} \right\rangle
= \frac{\mu_0 \omega^4 p_0^2}{32\pi^2 c}\sin^2\theta.
$$

Integrating over solid angle and using

$$
\int \sin^2\theta\,d\Omega = \frac{8\pi}{3},
$$

one finds the average total dipole power,

$$
\langle P \rangle = \frac{\mu_0 \omega^4 p_0^2}{12\pi c}.
$$

For a nonrelativistic point charge $q$ with acceleration $\mathbf a$, the dipole moment of the source is $\mathbf p = q\mathbf x$, so $\ddot{\mathbf p}=q\mathbf a$. One thereby obtains the nonrelativistic Larmor formula,

$$
P=\frac{\mu_0 q^2 a^2}{6\pi c} = \frac{q^2 a^2}{6\pi\varepsilon_0 c^3}.
$$

Its meaning is direct and profound: a charge radiates in proportion to the square of its acceleration. Velocity alone does not suffice.

## Hertzian Dipole: Canonical Elementary Source

The Hertzian dipole, or short oscillating dipole, is the standard elementary radiation source in classical electrodynamics and antenna theory. One imagines a thin current element of length $\ell$ carrying a time-harmonic current

$$
I(t)=I_0 \cos \omega t,
$$

with $\ell \ll \lambda$. In this limit the current is effectively uniform along the element and the source behaves as an electric dipole with moment amplitude

$$
p_0 = \frac{I_0 \ell}{\omega}.
$$

The far-zone fields are therefore

$$
E_\theta = \frac{\eta_0 k I_0 \ell}{4\pi r}\sin\theta\cos\!\left[\omega\left(t-\frac{r}{c}\right)\right],
\qquad
B_\phi = \frac{1}{c}E_\theta,
$$

where $k=\omega/c$ and $\eta_0=\sqrt{\mu_0/\varepsilon_0}=\mu_0 c$ is the impedance of free space. The angular dependence is the familiar toroidal pattern proportional to $\sin^2\theta$ in intensity.

This source is canonical because any sufficiently small localized radiator reduces, to leading order, to some combination of low multipoles. The short dipole is the simplest realization of electric dipole radiation and serves as the asymptotic model for a broad range of physical systems.

## Multipole Viewpoint

Electric dipole radiation is usually the leading contribution because it is the lowest multipole moment capable of time-dependent radiation for a localized neutral system. However, symmetry may force the electric dipole moment to vanish. Examples include sources with inversion symmetry or transitions between states for which the electric dipole matrix element is zero.

When electric dipole radiation vanishes, the next contributions are typically:

- **magnetic dipole radiation**, associated with time-dependent current loops or angular momentum structure,
- **electric quadrupole radiation**, associated with time-dependent second-rank charge moments.

The multipole expansion of radiation theory is therefore not merely classificatory; it is an asymptotic ordering principle. Higher multipoles are typically suppressed by powers of $ka$, where $a$ is the source size, but they become indispensable whenever lower moments vanish by symmetry or conservation laws.

## Radiation Reaction and Conceptual Remarks

A radiating charge loses energy and momentum through the fields it emits. Energy conservation therefore demands that work be done on the charge, or on the source system more generally, to sustain radiation. At a macroscopic level this is encoded by Poynting's theorem: the decrease of mechanical plus field energy inside a volume equals the outward flux of electromagnetic energy through the boundary.

At a more delicate level one encounters **radiation reaction**, namely the effect of the emitted field back on the motion of the source itself. The full treatment leads to subtleties such as self-force, runaway solutions, and pre-acceleration in the Abraham-Lorentz theory. Those issues lie beyond the present notebook, but the central conceptual point should be retained: radiation is not an externally appended phenomenon. It is part of the self-consistent energy accounting of Maxwell's theory.

## Canonical Worked Examples

### 1. Oscillating Electric Dipole

Let $\mathbf p(t)=p_0\cos \omega t\,\hat{\mathbf z}$. Then

$$
\ddot{\mathbf p}(t)=-\omega^2 p_0\cos \omega t\,\hat{\mathbf z}.
$$

Substituting into the dipole radiation formula gives a far-zone electric field proportional to $\omega^2 p_0 \sin\theta/r$. The two key structural facts are immediate: radiation strengthens with increasing temporal variation, and the angular dependence suppresses emission along the dipole axis.

### 2. Angular Radiation Pattern

The time-averaged intensity of electric dipole radiation is proportional to $\sin^2\theta$. Hence the emission is maximal in the plane perpendicular to the dipole axis and vanishes at $\theta=0$ and $\theta=\pi$. The resulting pattern is a torus in three dimensions and a figure-eight in a polar cross-section.

### 3. Power Radiated by a Nonrelativistic Accelerating Charge

For a single point charge with acceleration $\mathbf a$, one identifies $\ddot{\mathbf p}=q\mathbf a$ and obtains the Larmor formula

$$
P=\frac{q^2 a^2}{6\pi\varepsilon_0 c^3}.
$$

Thus a uniformly accelerated magnitude of acceleration yields constant radiated power in the nonrelativistic approximation. The result is quadratic in $q$ and in $a$.

### 4. Comparison of Near-Field and Far-Field Scaling

For a localized oscillating dipole, the full field contains terms scaling as $1/r^3$, $1/r^2$, and $1/r$. At small distance, the quasi-static $1/r^3$ contribution dominates. At intermediate distance, induction effects can be significant. At sufficiently large distance, only the $1/r$ term survives in the sense relevant for energy transport to infinity.

### 5. Symmetry and Allowed or Forbidden Dipole Radiation

If the source has a time-dependent charge distribution but its dipole moment remains identically zero by symmetry, then electric dipole radiation is absent. Radiation may still occur through magnetic dipole or electric quadrupole terms. This is one of the main practical uses of multipole analysis: it tells us not only what is present, but also what symmetry forbids.

## Computational Section

The following computations are not substitutes for the analysis above; they are visual complements to it. Each code cell is preceded by a short mathematical explanation and followed by a brief interpretation. The numerical work focuses on asymptotic structure rather than on a fully resolved solution of Maxwell's equations for an extended source.

We adopt a consistent plotting style and collect the physical constants needed for SI-unit illustrations.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 12
plt.rcParams['axes.titlesize'] = 13
plt.rcParams['axes.labelsize'] = 12

epsilon_0 = 8.8541878128e-12
mu_0 = 4.0e-7 * np.pi
c = 1.0 / np.sqrt(mu_0 * epsilon_0)
eta_0 = np.sqrt(mu_0 / epsilon_0)

print(f'c = {c:.6e} m/s')
print(f'eta_0 = {eta_0:.6f} ohms')
```

### Angular Radiation Pattern of an Oscillating Dipole

For a monochromatic electric dipole aligned with $\hat{\mathbf z}$, the time-averaged radiated power per unit solid angle is proportional to $\sin^2\theta$. The first plot shows this function directly, while the second gives the corresponding polar representation. This is the canonical dipole pattern.

```python
# Angular dependence of dipole radiation intensity.
theta = np.linspace(0.0, np.pi, 600)
theta_full = np.linspace(0.0, 2.0 * np.pi, 1000)

intensity = np.sin(theta) ** 2
intensity_full = np.sin(theta_full) ** 2

fig = plt.figure(figsize=(12, 4.8))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122, projection='polar')

ax1.plot(theta, intensity, color='darkred', linewidth=2)
ax1.set_xlabel(r'$\theta$')
ax1.set_ylabel(r'Normalized intensity $\propto \sin^2\theta$')
ax1.set_title('Dipole Radiation Pattern in Cartesian Form')

ax2.plot(theta_full, intensity_full, color='navy', linewidth=2)
ax2.set_title('Polar Plot of Dipole Radiation Pattern', va='bottom')

plt.tight_layout()
plt.show()
```

The vanishing intensity at $\theta=0$ and $\theta=\pi$ is a geometric consequence of transversality: along the dipole axis the acceleration projected perpendicular to the line of sight vanishes. The polar plot emphasizes the toroidal three-dimensional structure encoded in the two-dimensional cross-section.

### Near-, Induction-, and Radiation-Zone Scaling

A localized time-dependent dipole field contains terms with radial dependence $1/r^3$, $1/r^2$, and $1/r$. The precise coefficients depend on the source and on phase, but the hierarchy of powers is universal in the asymptotic expansion. The next computation compares these magnitudes on the same axes.

```python
# Compare the characteristic radial decay laws of different field zones.
r = np.linspace(0.2, 8.0, 800)
near = 1.0 / r**3
induction = 1.0 / r**2
radiation = 1.0 / r

fig, ax = plt.subplots()
ax.loglog(r, near, label=r'$1/r^3$ near field', linewidth=2)
ax.loglog(r, induction, label=r'$1/r^2$ induction field', linewidth=2)
ax.loglog(r, radiation, label=r'$1/r$ radiation field', linewidth=2)
ax.set_xlabel(r'$r$ (arbitrary units)')
ax.set_ylabel('Relative magnitude')
ax.set_title('Radial Scaling of Field Contributions')
ax.legend()
plt.show()
```

On logarithmic axes the three slopes are distinct and easy to compare. The graph should be read asymptotically: as $r$ increases, the $1/r$ term eventually dominates all faster-decaying contributions. This is why only the far-zone field is relevant for the total radiated power measured at large distance.

### Far-Zone Electric and Magnetic Field Amplitudes

For an oscillating dipole $\mathbf p(t)=p_0\cos \omega t\,\hat{\mathbf z}$, the peak far-zone amplitudes are

$$
E_{\theta,0}(r,\theta)=\frac{\mu_0 \omega^2 p_0}{4\pi r}\sin\theta,
\qquad
B_{\phi,0}(r,\theta)=\frac{E_{\theta,0}}{c}.
$$

The next code cell evaluates these amplitudes at fixed radius and then computes the corresponding cycle-averaged Poynting flux.

```python
# Far-zone amplitudes for a representative oscillating dipole.
p0 = 1.0e-29       # C m
frequency = 1.0e9  # Hz
omega = 2.0 * np.pi * frequency
r_obs = 5.0        # m

theta = np.linspace(0.0, np.pi, 500)
E0 = mu_0 * omega**2 * p0 * np.sin(theta) / (4.0 * np.pi * r_obs)
B0 = E0 / c
S_avg = E0**2 / (2.0 * mu_0 * c)

fig, axes = plt.subplots(1, 2, figsize=(12, 4.8))
axes[0].plot(theta, E0, label=r'$E_{\theta,0}$', linewidth=2)
axes[0].plot(theta, B0, label=r'$B_{\phi,0}$', linewidth=2)
axes[0].set_xlabel(r'$\theta$')
axes[0].set_ylabel('Amplitude (SI units)')
axes[0].set_title('Far-Zone Field Amplitudes')
axes[0].legend()

axes[1].plot(theta, S_avg, color='darkgreen', linewidth=2)
axes[1].set_xlabel(r'$\theta$')
axes[1].set_ylabel(r'$\langle S_r \rangle$ (W/m$^2$)')
axes[1].set_title('Cycle-Averaged Poynting Flux')

plt.tight_layout()
plt.show()
```

The electric and magnetic amplitudes share the same angular dependence because the far-zone fields are locked into the transverse radiation relation $\mathbf B = \hat{\mathbf r}\times \mathbf E/c$. Since the Poynting flux is quadratic in the field amplitude, its angular profile is proportional to $\sin^2\theta$, in agreement with the analytic derivation.

### Total Radiated Power as a Function of Frequency and Dipole Strength

For a monochromatic electric dipole, the average radiated power scales as $\langle P \rangle \propto \omega^4 p_0^2$. The quartic dependence on frequency is particularly important: even modest changes in frequency can greatly alter the radiated power. The following plots illustrate these parameter dependences in a simplified model.

```python
# Dependence of dipole power on frequency and dipole moment amplitude.
frequencies = np.logspace(6, 11, 300)
dipole_moments = np.logspace(-31, -27, 300)

p0_ref = 1.0e-29
omega_vals = 2.0 * np.pi * frequencies
power_vs_frequency = mu_0 * omega_vals**4 * p0_ref**2 / (12.0 * np.pi * c)

omega_ref = 2.0 * np.pi * 1.0e9
power_vs_p0 = mu_0 * omega_ref**4 * dipole_moments**2 / (12.0 * np.pi * c)

fig, axes = plt.subplots(1, 2, figsize=(12, 4.8))
axes[0].loglog(frequencies, power_vs_frequency, linewidth=2)
axes[0].set_xlabel('Frequency (Hz)')
axes[0].set_ylabel('Average power (W)')
axes[0].set_title(r'$\langle P \rangle$ versus Frequency at Fixed $p_0$')

axes[1].loglog(dipole_moments, power_vs_p0, linewidth=2, color='darkorange')
axes[1].set_xlabel(r'Dipole amplitude $p_0$ (C m)')
axes[1].set_ylabel('Average power (W)')
axes[1].set_title(r'$\langle P \rangle$ versus $p_0$ at Fixed Frequency')

plt.tight_layout()
plt.show()
```

The slopes on the log-log plots encode the analytic scaling laws directly. The first graph has slope $4$ because dipole radiation is controlled by the second time derivative of the dipole moment, and the second graph has slope $2$ because the radiated power is quadratic in field amplitude.

### Two-Dimensional Cross-Section of the Far-Zone Intensity

Although radiation is a three-dimensional phenomenon, it is often useful to examine a planar cross-section. For a dipole aligned with the vertical axis, the far-zone intensity in a meridional plane is proportional to $\sin^2\theta/r^2$. The next plot visualizes this quantity away from the origin, where the far-zone approximation is not intended to be trusted.

```python
# Cross-sectional map of far-zone intensity for a z-directed dipole.
grid = np.linspace(-4.0, 4.0, 401)
X, Z = np.meshgrid(grid, grid)
R = np.sqrt(X**2 + Z**2)
R_safe = np.maximum(R, 0.35)
sin_theta_sq = X**2 / R_safe**2
intensity_map = sin_theta_sq / R_safe**2

# Mask the immediate source region because the far-zone formula is not valid there.
mask = R < 0.5
intensity_map = np.ma.array(intensity_map, mask=mask)

fig, ax = plt.subplots(figsize=(6.5, 6.0))
positive_values = intensity_map.compressed()
positive_values = positive_values[positive_values > 0.0]
levels = np.geomspace(positive_values.min(), positive_values.max(), 24)
contour = ax.contourf(X, Z, intensity_map, levels=levels, cmap='inferno')
ax.set_xlabel('$x$')
ax.set_ylabel('$z$')
ax.set_title('Cross-Section of Dipole Far-Zone Intensity')
ax.set_aspect('equal')
fig.colorbar(contour, ax=ax, label='Relative intensity')
plt.show()
```

The dark lobes along the horizontal direction correspond to maximal radiation in the plane orthogonal to the dipole axis, while the nulls along the vertical axis are the familiar dipole zeros. The excluded central region is important conceptually: the far-zone formula is asymptotic and should not be interpreted as an exact near-source field.

### Simple Animation of an Oscillating Dipole Moment and Its Pattern

The final computation animates an oscillating dipole moment together with the instantaneous angular weighting $\sin^2\theta\,\cos^2\omega t$. This is not a full Maxwell solver; rather, it is a pedagogical visualization of how a time-dependent dipole moment drives an oscillatory radiation pattern whose spatial shape is fixed while the instantaneous intensity pulsates.

```python
# Animate a simple dipole oscillation and the corresponding instantaneous intensity pattern.
phi = np.linspace(0.0, 2.0 * np.pi, 600)
base_pattern = np.sin(phi) ** 2
times = np.linspace(0.0, 2.0 * np.pi, 80)

fig = plt.figure(figsize=(11, 4.8))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122, projection='polar')

dipole_line, = ax1.plot([], [], color='darkred', linewidth=4)
ax1.set_xlim(-0.2, 0.2)
ax1.set_ylim(-1.2, 1.2)
ax1.set_xticks([])
ax1.set_ylabel('Dipole amplitude')
ax1.set_title('Oscillating Dipole Moment')

pattern_line, = ax2.plot([], [], color='navy', linewidth=2)
ax2.set_title('Instantaneous Angular Weighting', va='bottom')

def init():
    dipole_line.set_data([0.0, 0.0], [0.0, 0.0])
    pattern_line.set_data(phi, np.zeros_like(phi))
    return dipole_line, pattern_line

def update(frame):
    amplitude = np.cos(frame)
    dipole_line.set_data([0.0, 0.0], [-amplitude, amplitude])
    pattern_line.set_data(phi, base_pattern * amplitude**2)
    return dipole_line, pattern_line

anim = FuncAnimation(fig, update, frames=times, init_func=init, interval=100, blit=True)
plt.close(fig)
HTML(anim.to_jshtml())
```

The animation makes two points transparent. First, reversing the sign of the dipole moment does not change the radiated intensity because the power is quadratic in the field. Second, the angular structure remains dipolar throughout the oscillation: the source strength varies in time, but the symmetry of the radiation pattern is fixed by the source geometry.

## Advanced Conceptual Section

A uniformly moving charge does not radiate in the same way as an accelerated charge because its field configuration, although time-dependent in a given frame, does not contain the same asymptotic transverse $1/r$ component associated with irreversible energy transport to infinity. Acceleration changes the field in a way that cannot be accounted for by mere translation of a static configuration; it generates genuine disturbances that detach from the source and propagate outward.

Retarded time is the mathematical expression of causality in this context. The field at a distant point is determined by the source history on the past light cone, not by the source at the same coordinate time. This is why the wave equation and the retarded Green's function are so central: they encode both the propagation speed and the correct causal support of the solution.

The far field carries energy to infinity precisely because its amplitude scales as $1/r$. The corresponding Poynting vector therefore scales as $1/r^2$, so the flux through a sphere of area $4\pi r^2$ tends to a finite limit. Faster-decaying fields do not survive this large-radius integration and therefore cannot represent outgoing radiation in the strict asymptotic sense.

## Key Takeaways

- Radiation is the part of the electromagnetic field that propagates causally and carries finite energy to infinity.
- The Lorenz gauge converts the potential equations into inhomogeneous wave equations with retarded solutions.
- Retarded potentials make causality explicit: fields depend on earlier source values located on the past light cone.
- In the far zone, the fields are transverse, scale as $1/r$, and satisfy $\mathbf B = \hat{\mathbf r}\times \mathbf E / c$.
- Electric dipole radiation is the leading radiative contribution for many localized sources and has angular intensity proportional to $\sin^2\theta$.
- The average dipole power scales as $\omega^4 p_0^2$, and for a point charge reduces to the Larmor formula.

## Common Pitfalls

- Confusing time-dependent near fields with radiation fields. Not every time-varying field transports power to infinity.
- Replacing retarded source values by instantaneous ones. This destroys the causal structure of the theory.
- Forgetting that the far-zone approximation applies asymptotically and should not be used near the source.
- Treating dipole radiation as universal without checking whether the electric dipole moment vanishes by symmetry.
- Ignoring the distinction between field amplitude scaling and power scaling.

## Connection to Later Topics

The present development leads directly to retarded Green's functions, systematic multipole radiation expansions, relativistic radiation from moving charges, antenna theory, scattering, and synchrotron or bremsstrahlung-type processes. In each of these subjects the same structural ideas reappear: wave equations, causality, asymptotic fields, source moments, and energy flux.
