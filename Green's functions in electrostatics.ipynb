# Green's Functions in Electrostatics

Green's functions provide one of the most systematic ways to solve Poisson and Laplace boundary value problems in electrostatics. They convert a differential equation into an integral representation in which the geometry and the boundary conditions are built directly into the kernel. In Jackson-style electrodynamics, this is not merely a formal trick: it is a unifying framework that clarifies the role of the delta function, the meaning of the Coulomb potential, the logic of the method of images, and the structure of electrostatic boundary value theory.

## Motivation and Context

Electrostatic potentials satisfy Poisson's equation,

$$
\nabla^2 \phi(\mathbf{r}) = -\frac{\rho(\mathbf{r})}{\varepsilon_0}.
$$

In free space, this equation admits the familiar Coulomb integral solution. In the presence of boundaries, however, the problem becomes substantially more subtle. One may know the potential on a conducting surface, the normal electric field on a boundary, or some mixed set of conditions. Direct solution of the differential equation then depends strongly on the geometry.

Green's functions provide a systematic response kernel adapted to the operator and to the boundary conditions. Roughly speaking, the Green's function is the potential generated at the field point $\mathbf{r}$ by a unit point source placed at the source point $\mathbf{r}'$, subject to the same boundary conditions as the physical problem.

This viewpoint turns the electrostatic problem into a superposition principle of the most refined kind: every source distribution is represented as a continuous superposition of point sources, and the boundary geometry is incorporated from the beginning rather than appended afterward.

## Mathematical Definition

Throughout this notebook we adopt the convention

$$
\nabla^2 G(\mathbf{r},\mathbf{r}') = -4\pi\delta^{(3)}(\mathbf{r}-\mathbf{r}').
$$

This is the conventional electrostatic normalization used in Jackson. The SI factor $1/(4\pi\varepsilon_0)$ then appears in the potential representation formula rather than inside the definition of the Green's function itself.

The differential operator acts on the field point $\mathbf{r}$, while the source point $\mathbf{r}'$ is held fixed. This distinction is essential:

- $\mathbf{r}'$ locates the point source.
- $\mathbf{r}$ is the point at which the potential is evaluated.

Thus, for fixed $\mathbf{r}'$, the function $G(\mathbf{r},\mathbf{r}')$ is harmonic everywhere in $\mathbf{r}$ except at $\mathbf{r}=\mathbf{r}'$, where it has the singularity required to reproduce a delta distribution.

Under suitable conditions, particularly for self-adjoint boundary value problems, one has the reciprocity relation

$$
G(\mathbf{r},\mathbf{r}') = G(\mathbf{r}',\mathbf{r}).
$$

This symmetry is not automatic for every conceivable formulation, but it holds in the standard electrostatic Dirichlet and suitably normalized Neumann settings.

## Free-Space Green's Function

In unbounded free space, the Green's function is determined by translational and rotational invariance. It can depend only on the separation $|\mathbf{r}-\mathbf{r}'|$, and the appropriate solution is

$$
G_0(\mathbf{r},\mathbf{r}') = \frac{1}{|\mathbf{r}-\mathbf{r}'|}.
$$

Away from the singular point, one has

$$
\nabla^2 \frac{1}{|\mathbf{r}-\mathbf{r}'|} = 0,
\qquad \mathbf{r}\neq \mathbf{r}'.
$$

Globally, however, the correct distributional identity is

$$
\nabla^2 \frac{1}{|\mathbf{r}-\mathbf{r}'|} = -4\pi\delta^{(3)}(\mathbf{r}-\mathbf{r}').
$$

This is the mathematical statement that the Coulomb kernel is the fundamental solution of the Laplacian in three dimensions.

The Coulomb potential for a point charge $q$ at $\mathbf{r}'$ follows immediately:

$$
\phi(\mathbf{r}) = \frac{q}{4\pi\varepsilon_0} G_0(\mathbf{r},\mathbf{r}')
= \frac{1}{4\pi\varepsilon_0}\frac{q}{|\mathbf{r}-\mathbf{r}'|}.
$$

For a general charge density, superposition yields the free-space solution of Poisson's equation,

$$
\phi(\mathbf{r}) = \frac{1}{4\pi\varepsilon_0}\int \frac{\rho(\mathbf{r}')}{|\mathbf{r}-\mathbf{r}'|}\,d^3r'.
$$

Thus the usual Coulomb integral is nothing more than the Green's-function solution with the free-space kernel.

## Derivation of the Green's Function Solution

Let $\phi(\mathbf{r}')$ satisfy Poisson's equation in a region $V$,

$$
\nabla'^2 \phi(\mathbf{r}') = -\frac{\rho(\mathbf{r}')}{\varepsilon_0},
$$

and let $G(\mathbf{r},\mathbf{r}')$ satisfy

$$
\nabla'^2 G(\mathbf{r},\mathbf{r}') = -4\pi\delta^{(3)}(\mathbf{r}'-\mathbf{r}).
$$

Applying Green's second identity in the source variable $\mathbf{r}'$,

$$
\int_V \left(\phi\nabla'^2 G - G\nabla'^2\phi\right)d^3r'
= \oint_S \left(\phi\frac{\partial G}{\partial n'} - G\frac{\partial \phi}{\partial n'}\right)dA',
$$

and substituting the defining equations gives

$$
-4\pi \phi(\mathbf{r}) + \frac{1}{\varepsilon_0}\int_V G(\mathbf{r},\mathbf{r}')\rho(\mathbf{r}')\,d^3r'
= \oint_S \left(\phi\frac{\partial G}{\partial n'} - G\frac{\partial \phi}{\partial n'}\right)dA'.
$$

Rearranging,

$$
\phi(\mathbf{r})
= \frac{1}{4\pi\varepsilon_0}\int_V \rho(\mathbf{r}')G(\mathbf{r},\mathbf{r}')\,d^3r'
- \frac{1}{4\pi}\oint_S \phi(\mathbf{r}')\frac{\partial G}{\partial n'}\,dA'
+ \frac{1}{4\pi}\oint_S G(\mathbf{r},\mathbf{r}')\frac{\partial \phi}{\partial n'}\,dA'.
$$

This is the general Green's-function representation formula.

The first term is the volume contribution from the source distribution. The boundary terms encode the influence of the enclosing surfaces. Their presence is essential: in bounded electrostatic problems, the solution is determined not only by charges in the volume but also by the boundary data.

The power of the method lies in choosing a Green's function adapted to the specified boundary condition so that one of the surface terms vanishes or simplifies.

## Dirichlet and Neumann Green's Functions

### Dirichlet Green's Function

A Dirichlet Green's function is chosen to satisfy

$$
G_D(\mathbf{r},\mathbf{r}') = 0 \qquad \text{for } \mathbf{r}' \in S.
$$

Then the representation formula becomes

$$
\phi(\mathbf{r})
= \frac{1}{4\pi\varepsilon_0}\int_V \rho(\mathbf{r}')G_D(\mathbf{r},\mathbf{r}')\,d^3r'
- \frac{1}{4\pi}\oint_S \phi(\mathbf{r}')\frac{\partial G_D}{\partial n'}\,dA'.
$$

This is the natural choice when the boundary potential is specified. Physically, the Green's function is the point-source response in a region whose boundary is held at zero potential.

### Neumann Green's Function

A Neumann Green's function is adapted to prescribed normal derivative on the boundary. One chooses

$$
\frac{\partial G_N}{\partial n'} = \text{specified function on } S,
$$

typically arranged so that the boundary term involving $\partial \phi/\partial n'$ becomes the directly usable one. In the pure Neumann problem, the potential is determined only up to an additive constant. Correspondingly, the Green's function is defined only up to the addition of functions consistent with that ambiguity, together with an appropriate compatibility condition on the total normal derivative.

The constant ambiguity is not a defect. It reflects the physical fact that only potential differences matter unless a reference level is fixed.

## Reciprocity and Symmetry

Suppose $G(\mathbf{r},\mathbf{r}')$ and $G(\mathbf{r}',\mathbf{r})$ satisfy the same self-adjoint boundary value problem. Applying Green's identity to the pair of functions

$$
u(\mathbf{x}) = G(\mathbf{x},\mathbf{r}),
\qquad
v(\mathbf{x}) = G(\mathbf{x},\mathbf{r}'),
$$

one finds, after the boundary terms vanish appropriately,

$$
G(\mathbf{r},\mathbf{r}') = G(\mathbf{r}',\mathbf{r}).
$$

This reciprocity relation relies on the self-adjoint character of the Laplacian together with compatible boundary conditions. Mathematically, it expresses the symmetry of the inverse operator. Physically, it says that the response measured at $\mathbf{r}$ due to a source at $\mathbf{r}'$ is the same as the response measured at $\mathbf{r}'$ due to an identical source placed at $\mathbf{r}$, provided the surrounding boundary conditions are the same.

## Canonical Examples

### 1. Free-Space Green's Function

The fundamental solution in unbounded space is

$$
G_0(\mathbf{r},\mathbf{r}') = \frac{1}{|\mathbf{r}-\mathbf{r}'|}.
$$

It is translationally invariant and depends only on the distance between source and field points.

### 2. Dirichlet Green's Function for a Grounded Plane

Consider the half-space $z>0$ with boundary plane $z=0$ held at zero potential. A suitable Dirichlet Green's function is

$$
G_D(\mathbf{r},\mathbf{r}')
= \frac{1}{|\mathbf{r}-\mathbf{r}'|} - \frac{1}{|\mathbf{r}-\mathbf{r}'_{\mathrm{image}}|},
$$

where the image point is obtained by reflecting $\mathbf{r}'=(x',y',z')$ across the plane:

$$
\mathbf{r}'_{\mathrm{image}}=(x',y',-z').
$$

On the boundary $z=0$, the two distances are equal, so $G_D=0$. Thus the boundary condition is built directly into the kernel.

### 3. Green's Function Inside a Spherical Boundary

For a sphere of radius $a$ with grounded boundary, one can construct a Dirichlet Green's function by combining the free-space singularity with an image term placed at the Kelvin-inverted point. The result is more intricate than the planar case, but the logic is identical: the singular Coulomb kernel is corrected by a harmonic term chosen to enforce the boundary condition on the sphere.

These examples show the general strategy: the Green's function consists of the fundamental singular piece plus an additional harmonic contribution selected so that the required boundary condition is satisfied.

## Relationship to the Method of Images

The method of images is best understood as a special technique for constructing Green's functions in geometries with enough symmetry.

For the grounded conducting plane, the Green's function is

$$
G(\mathbf{r},\mathbf{r}') = \frac{1}{|\mathbf{r}-\mathbf{r}'|} - \frac{1}{|\mathbf{r}-\mathbf{r}'_{\text{image}}|}.
$$

The first term gives the singular response to the source itself. The second term is harmonic in the physical half-space and is chosen precisely so that the total Green's function vanishes on the plane. This construction is exactly what the method of images produces.

Thus the method of images is not a separate theory. It is a particularly elegant way of building a Green's function for certain special boundaries where a reflected or transformed source can enforce the boundary condition.

## Boundary Value Problem Perspective

A Green's function encodes two pieces of information simultaneously:

- the inverse of the differential operator,
- and the boundary condition appropriate to the physical problem.

This is what makes Green's functions so powerful in electrostatics. Once the correct kernel is found, the potential follows by an integral formula, and the uniqueness theorem ensures that the result is the physical solution. In effect, all the geometric difficulty of the boundary value problem is shifted into the construction of the kernel itself.

## Computational Section

The computational examples below are designed to make the Green's-function viewpoint concrete. We will visualize the free-space kernel, compare it with the grounded-plane Green's function, and use the image construction to show how the same source produces very different potentials in the presence of a boundary.

For clarity, we work in dimensionless units and suppress the overall factor $1/(4\pi\varepsilon_0)$ in most plots. The geometry of the Green's function is the main focus.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 12
plt.rcParams['axes.titlesize'] = 13
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 10


def free_space_green_2d_slice(x, z, xp, zp, epsilon=1e-6):
    """Three-dimensional free-space Green's function restricted to the y=0 plane."""
    r = np.sqrt((x - xp) ** 2 + (z - zp) ** 2)
    return 1.0 / np.maximum(r, epsilon)


def grounded_plane_green_2d_slice(x, z, xp, zp, epsilon=1e-6):
    """Dirichlet Green's function for the half-space z>0 using the image construction."""
    r_source = np.sqrt((x - xp) ** 2 + (z - zp) ** 2)
    r_image = np.sqrt((x - xp) ** 2 + (z + zp) ** 2)
    return 1.0 / np.maximum(r_source, epsilon) - 1.0 / np.maximum(r_image, epsilon)


def point_charge_potential_grounded_plane(x, z, q=1.0, xp=0.0, zp=1.0, epsilon=1e-6):
    """Potential for a point charge above a grounded plane, up to 1/(4 pi epsilon0)."""
    return q * grounded_plane_green_2d_slice(x, z, xp, zp, epsilon=epsilon)


def numerical_convolution_1d(x_obs, x_src, rho, kernel_epsilon=0.02):
    """A regularized 1D analogue of a Green-function convolution for pedagogical plotting."""
    dx = x_src[1] - x_src[0]
    phi = np.zeros_like(x_obs)
    for i, xo in enumerate(x_obs):
        kernel = 1.0 / np.sqrt((xo - x_src) ** 2 + kernel_epsilon**2)
        phi[i] = np.sum(rho * kernel) * dx
    return phi
```

### Free-Space Green's Function as a Spatial Response Kernel

We first visualize the free-space Green's function on a two-dimensional slice through three-dimensional space. The kernel

$$
G_0(\mathbf{r},\mathbf{r}') = \frac{1}{|\mathbf{r}-\mathbf{r}'|}
$$

is singular at the source point and decays away from it. In the plot below, the source is placed at a fixed point above the origin in the $(x,z)$ plane.

```python
x = np.linspace(-2.5, 2.5, 300)
z = np.linspace(0.0, 3.0, 260)
X, Z = np.meshgrid(x, z, indexing='xy')

xp, zp = 0.0, 1.0
G_free = free_space_green_2d_slice(X, Z, xp, zp, epsilon=0.03)

fig, ax = plt.subplots(figsize=(7, 5.5))
im = ax.imshow(
    G_free,
    extent=[x.min(), x.max(), z.min(), z.max()],
    origin='lower',
    cmap='viridis',
    aspect='auto'
)
ax.scatter([xp], [zp], color='red', s=40, label='Source point')
ax.set_title('Free-Space Green\'s Function on an $(x,z)$ Slice')
ax.set_xlabel('$x$')
ax.set_ylabel('$z$')
ax.legend(loc='upper right')
fig.colorbar(im, ax=ax, label=r'$G_0$')
plt.show()
```

The large peak marks the singularity at the source point. Away from that point, the kernel is smooth and decreases with distance. This is the universal free-space response of the Laplacian in three dimensions.

### 1D Slice Through the Singularity Structure

A line slice is useful for visualizing the singular behavior more transparently. We compare a horizontal slice through the source height with a slice well away from the source. The singularity appears only when the observation line passes through the source point.

```python
x_line = np.linspace(-2.0, 2.0, 1200)
slice_at_source = free_space_green_2d_slice(x_line, zp * np.ones_like(x_line), xp, zp, epsilon=0.02)
slice_above = free_space_green_2d_slice(x_line, (zp + 0.6) * np.ones_like(x_line), xp, zp, epsilon=0.02)

fig, ax = plt.subplots()
ax.plot(x_line, slice_at_source, label=r'$z=z_0$')
ax.plot(x_line, slice_above, label=r'$z=z_0+0.6$')
ax.set_title('Line Slices of the Free-Space Green\'s Function')
ax.set_xlabel('$x$')
ax.set_ylabel(r'$G_0$')
ax.set_ylim(0, 18)
ax.legend()
plt.show()
```

The slice through the source exhibits the regularized spike associated with the Coulomb singularity, while the displaced slice remains finite everywhere. This highlights the distinction between the singular source point and the harmonic behavior away from it.

### Green's Function for a Grounded Plane

For the half-space $z>0$ with a grounded plane at $z=0$, the Dirichlet Green's function is obtained by subtracting the image contribution. The resulting kernel vanishes on the boundary while preserving the correct point singularity in the physical region.

```python
G_plane = grounded_plane_green_2d_slice(X, Z, xp, zp, epsilon=0.03)

fig, ax = plt.subplots(figsize=(7, 5.5))
im = ax.imshow(
    G_plane,
    extent=[x.min(), x.max(), z.min(), z.max()],
    origin='lower',
    cmap='magma',
    aspect='auto'
)
ax.scatter([xp], [zp], color='cyan', s=40, label='Source point')
ax.axhline(0.0, color='white', linewidth=2, label='Grounded plane')
ax.set_title('Dirichlet Green\'s Function for a Grounded Plane')
ax.set_xlabel('$x$')
ax.set_ylabel('$z$')
ax.legend(loc='upper right')
fig.colorbar(im, ax=ax, label=r'$G_D$')
plt.show()
```

Compared with the free-space kernel, the response is suppressed near the boundary because the image term enforces zero potential at the plane. The source singularity remains, but the surrounding field is reshaped globally by the boundary condition.

### Potential of a Point Charge Near a Grounded Plane

For a point charge above a grounded conducting plane, the physical potential in the half-space is proportional to the grounded-plane Green's function. Contour plots make the geometric effect of the image construction especially clear.

```python
phi_plane = point_charge_potential_grounded_plane(X, Z, q=1.0, xp=xp, zp=zp, epsilon=0.03)

fig, ax = plt.subplots(figsize=(7, 5.5))
levels = np.linspace(np.percentile(phi_plane, 5), np.percentile(phi_plane, 99), 25)
cont = ax.contourf(X, Z, phi_plane, levels=levels, cmap='viridis')
ax.contour(X, Z, phi_plane, levels=levels[::3], colors='white', linewidths=0.6)
ax.scatter([xp], [zp], color='red', s=40, label='Charge')
ax.axhline(0.0, color='black', linewidth=2, label='Grounded plane')
ax.set_title('Potential of a Point Charge Near a Grounded Plane')
ax.set_xlabel('$x$')
ax.set_ylabel('$z$')
ax.legend(loc='upper right')
fig.colorbar(cont, ax=ax, label=r'Potential up to $1/(4\pi\varepsilon_0)$')
plt.show()
```

The potential contours meet the grounded plane in a way consistent with zero potential there. Relative to free space, the field is compressed toward the boundary because the image charge contributes an equal and opposite response. This is the electrostatic content of the Dirichlet Green's function for the half-space.

### Free Space Versus Grounded Boundary for the Same Source

The same physical source produces different potentials in different boundary environments. We compare the free-space kernel and the grounded-plane kernel for the same source point to make this dependence explicit.

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4.8))

im0 = axes[0].imshow(G_free, extent=[x.min(), x.max(), z.min(), z.max()], origin='lower', cmap='viridis', aspect='auto')
axes[0].set_title('Free Space')
axes[0].set_xlabel('$x$')
axes[0].set_ylabel('$z$')
fig.colorbar(im0, ax=axes[0], shrink=0.85)

im1 = axes[1].imshow(G_plane, extent=[x.min(), x.max(), z.min(), z.max()], origin='lower', cmap='magma', aspect='auto')
axes[1].set_title('Grounded Plane')
axes[1].set_xlabel('$x$')
axes[1].set_ylabel('$z$')
fig.colorbar(im1, ax=axes[1], shrink=0.85)

im2 = axes[2].imshow((G_plane - G_free), extent=[x.min(), x.max(), z.min(), z.max()], origin='lower', cmap='coolwarm', aspect='auto')
axes[2].set_title('Boundary-Induced Modification')
axes[2].set_xlabel('$x$')
axes[2].set_ylabel('$z$')
fig.colorbar(im2, ax=axes[2], shrink=0.85)

plt.tight_layout()
plt.show()
```

The third panel isolates the boundary contribution. Mathematically, this is the added harmonic term that converts the fundamental solution into the Green's function appropriate to the half-space. Physically, it is the boundary-induced distortion of the response to the source.

### Numerical Evaluation of a Green-Function Integral in a Simplified Setting

To illustrate the integral-superposition viewpoint directly, we compute a regularized one-dimensional analogue of a Green-function convolution for a smooth localized charge distribution. The goal is pedagogical rather than literal: it makes transparent how a source distribution is turned into a potential by integrating against a kernel.

```python
x_src = np.linspace(-2.0, 2.0, 1200)
rho = np.exp(-((x_src - 0.45) / 0.35) ** 2) + 0.6 * np.exp(-((x_src + 0.8) / 0.25) ** 2)
phi_conv = numerical_convolution_1d(x_src, x_src, rho, kernel_epsilon=0.04)

fig, axes = plt.subplots(2, 1, figsize=(8, 8), sharex=True)

axes[0].plot(x_src, rho, color='tab:red')
axes[0].set_ylabel(r'$\rho(x)$')
axes[0].set_title('Regularized Green-Function Superposition in 1D Analogue')

axes[1].plot(x_src, phi_conv, color='tab:blue')
axes[1].set_ylabel(r'$\phi(x)$')
axes[1].set_xlabel('$x$')

plt.tight_layout()
plt.show()
```

Although this is a simplified analogue rather than the exact three-dimensional electrostatic integral, it captures the essential logic of the Green's-function method: the source is weighted by a response kernel and summed over all source points. The resulting potential is smoother and more spatially extended than the original charge distribution because the kernel spreads the influence of each source point throughout the domain.

## Advanced Conceptual Section

A Green's function is not unique until the boundary conditions are specified. The singular part is fixed locally by the requirement that the Laplacian reproduce the delta function, but one may always add a function of $\mathbf{r}$ that is harmonic in the domain away from the source point. Different choices correspond to different boundary value problems.

This leads to an important distinction:

- The **fundamental solution** is the free-space kernel $1/|\mathbf{r}-\mathbf{r}'|$.
- The **Green's function for a bounded domain** is the fundamental singularity plus whatever additional harmonic correction is needed to enforce the prescribed boundary condition.

From the operator viewpoint, the Green's function is the inverse of the differential operator in a distributional sense. More precisely, for the Laplacian subject to specified boundary conditions, the Green's function is the integral kernel of the inverse map from source to solution. This language becomes central far beyond electrostatics: wave equations, Helmholtz problems, diffusion equations, quantum mechanics, and field theory all rely on Green-function ideas in one form or another.

## Key Takeaways

- Green's functions convert Poisson and Laplace boundary value problems into integral representations.
- The Green's function is the response to a unit point source, with the boundary conditions of the physical problem built directly into the kernel.
- In the Jackson convention used here, the Green's function satisfies $\nabla^2 G = -4\pi\delta^{(3)}$.
- The free-space Green's function is $1/|\mathbf{r}-\mathbf{r}'|$, which is the fundamental solution of the Laplacian in three dimensions.
- Boundary-adapted Green's functions differ from the free-space kernel by added harmonic terms chosen to enforce Dirichlet or Neumann conditions.
- The method of images is a special way of constructing Green's functions in highly symmetric geometries.
- Reciprocity and uniqueness make Green's-function methods both mathematically rigorous and physically reliable.

## Common Pitfalls

- Confusing the free-space fundamental solution with the Green's function appropriate to a bounded domain.
- Forgetting which variable the Laplacian acts on in $G(\mathbf{r},\mathbf{r}')$.
- Neglecting the boundary terms in the representation formula when the Green's function has not been chosen to eliminate them.
- Treating Dirichlet and Neumann Green's functions as interchangeable even though they encode different physical data.
- Overlooking the additive constant ambiguity in Neumann problems.
- Using an image construction without verifying that the resulting kernel satisfies the correct boundary condition and singularity structure.

## Connection to Later Topics

Green's functions sit at the center of later electrodynamics and mathematical physics:

- **Method of images:** special Green-function constructions for simple boundaries.
- **Multipole expansion:** asymptotic analysis of Green-function integrals for localized sources.
- **Wave equations:** time-dependent analogues lead to retarded and advanced Green's functions.
- **Radiation theory:** propagators organize fields produced by time-varying sources.
- **Mathematical physics more broadly:** operator inversion, spectral theory, and integral equations all rely on the same conceptual framework.
