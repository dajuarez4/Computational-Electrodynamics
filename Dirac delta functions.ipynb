# Dirac Delta Function

This notebook develops the Dirac delta function at the level used in graduate electrodynamics, with the style and rigor associated with Jackson. The goal is twofold:

1. To understand the delta function as a mathematical distribution rather than an ordinary function.
2. To see why it is indispensable for compact descriptions of localized charge distributions in electrodynamics.

Throughout, we balance formal definitions with computational intuition using Gaussian approximations that become increasingly localized while preserving total weight.

## Motivation and Physical Meaning

In electrodynamics, one repeatedly encounters idealized sources that are sharply localized in space: a point charge, a thin charged wire, or an infinitesimally thin charged sheet. Physically, these are limiting descriptions of very concentrated charge distributions. Mathematically, ordinary functions are often inadequate for expressing such idealizations in a way that preserves the correct total charge.

The Dirac delta function solves this problem. It is written suggestively as though it were a function, but its real meaning is operational: it extracts the value of a test function at a point. This makes it the natural language for describing localized sources in Maxwell's equations and in Poisson's equation.

A useful physical picture is the following: a delta distribution represents a quantity concentrated into an infinitesimal region while keeping a finite integral. That is precisely what is needed when the total charge is finite but its spatial extent is idealized to zero thickness.

## Mathematical Definition of the Dirac Delta Function

The symbol $\delta(x-a)$ is not an ordinary function in the classical sense. If one tried to interpret it naively as being zero everywhere except at $x=a$, where it is somehow infinite, one would immediately leave the category of usual functions. The correct framework is that of **distributions** (or generalized functions).

A distribution is defined by how it acts under integration against sufficiently smooth test functions. The Dirac delta centered at $a$ is characterized by the **sifting property**

$$
\int_{-\infty}^{\infty} f(x)\,\delta(x-a)\,dx = f(a),
$$

for suitable test functions $f$. This equation is the definition that matters. It says that $\delta(x-a)$ extracts the value of a function at the point $x=a$.

It is also customary to write

$$
\delta(x-a)=0 \quad \text{for } x\neq a,
$$

and

$$
\int_{-\infty}^{\infty}\delta(x-a)\,dx = 1,
$$

but these statements are shorthand for the distributional meaning above. The delta function is therefore best thought of not as a usual graphable object, but as a linear functional whose effect is seen only inside an integral.

## Main Properties

### 1. Sifting Property

The defining property is

$$
\int_{-\infty}^{\infty} f(x)\delta(x-a)\,dx = f(a).
$$

This is the statement that makes the delta distribution useful. Inside an integral, it localizes the integrand at a single point.

### 2. Translation Property

The shift from $\delta(x)$ to $\delta(x-a)$ simply relocates the point at which the sampling occurs:

$$
\delta(x-a) \text{ samples at } x=a.
$$

Equivalently,

$$
\int_{-\infty}^{\infty} f(x)\delta(x-a)\,dx = f(a),
$$

so the translation moves the support of the distribution from the origin to $a$.

### 3. Scaling Property

A crucial identity is

$$
\delta(ax)=\frac{1}{|a|}\delta(x), \qquad a\neq 0.
$$

This follows from a change of variables in an integral. The factor $1/|a|$ ensures that the total weight remains unity:

$$
\int_{-\infty}^{\infty} f(x)\delta(ax)\,dx
= \int_{-\infty}^{\infty} f(x)\frac{1}{|a|}\delta(x)\,dx
= \frac{1}{|a|}f(0).
$$

More generally, if $g(x)$ has isolated simple zeros $x_i$, then

$$
\delta(g(x)) = \sum_i \frac{\delta(x-x_i)}{|g'(x_i)|}.
$$

This identity appears frequently in coordinate transformations and in electrodynamics problems with symmetry.

### 4. Gaussian Representation

A standard approximation to the delta distribution is a sequence of normalized Gaussians:

$$
\delta_\varepsilon(x)=\frac{1}{\sqrt{\pi}\,\varepsilon}e^{-x^2/\varepsilon^2}.
$$

Each $\delta_\varepsilon(x)$ is an ordinary function with unit area:

$$
\int_{-\infty}^{\infty}\delta_\varepsilon(x)\,dx = 1.
$$

As $\varepsilon \to 0$, the Gaussian becomes narrower and taller, and in the distributional sense

$$
\delta_\varepsilon(x) \to \delta(x).
$$

This limit is not pointwise in the ordinary function sense; it is meaningful only under integration against test functions.

## Delta Function in 1D, 2D, and 3D

### One Dimension

In one dimension, the delta distribution $\delta(x-a)$ localizes at a single point on the line. Its defining action is

$$
\int_{-\infty}^{\infty} f(x)\delta(x-a)\,dx = f(a).
$$

### Two Dimensions

In two dimensions, the delta distribution is written

$$
\delta^{(2)}(\mathbf{r}-\mathbf{r}_0)=\delta(x-x_0)\delta(y-y_0).
$$

Its defining property is

$$
\int\!\!\int f(x,y)\,\delta(x-x_0)\delta(y-y_0)\,dx\,dy = f(x_0,y_0).
$$

This describes localization at a point in a plane.

### Three Dimensions

In three-dimensional space,

$$
\delta^{(3)}(\mathbf{r}-\mathbf{r}_0)=\delta(x-x_0)\delta(y-y_0)\delta(z-z_0),
$$

and it satisfies

$$
\iiint f(\mathbf{r})\,\delta^{(3)}(\mathbf{r}-\mathbf{r}_0)\,d^3r = f(\mathbf{r}_0).
$$

It has dimensions of inverse volume. This dimensional point is important in electrodynamics: if $\rho(\mathbf{r})$ is a volume charge density, then $\delta^{(3)}(\mathbf{r}-\mathbf{r}_0)$ must scale as $1/\text{volume}$ so that the total charge comes out with the correct units.

In curvilinear coordinates one must also include the Jacobian. For example, in spherical coordinates,

$$
\delta^{(3)}(\mathbf{r}-\mathbf{r}_0)
= \frac{1}{r^2\sin\theta}\delta(r-r_0)\delta(\theta-\theta_0)\delta(\phi-\phi_0),
$$

with the understanding that the volume element is $d^3r=r^2\sin\theta\,dr\,d\theta\,d\phi$.

## Delta Function in Electrodynamics

The delta distribution appears naturally whenever charge is idealized as localized on sets of lower dimension.

For a point charge $q$ located at $\mathbf{r}_0$, the volume charge density is

$$
\rho(\mathbf{r}) = q\,\delta^{(3)}(\mathbf{r}-\mathbf{r}_0).
$$

This formula is exact in the distributional sense because

$$
\iiint \rho(\mathbf{r})\,d^3r = q\iiint \delta^{(3)}(\mathbf{r}-\mathbf{r}_0)\,d^3r = q.
$$

Similarly, if charge is concentrated along a line, the density can still be represented as a three-dimensional volume density by using delta functions in the transverse coordinates. If charge is concentrated on a surface, one uses a delta function normal to the surface.

This viewpoint is especially powerful because Maxwell's equations and Poisson's equation are written in terms of volume densities. Delta functions let lower-dimensional charge distributions be embedded into that common framework without changing the field equations.

## Examples with Point, Line, and Surface Charge Densities

### 1. Point Charge in 3D

A point charge $q$ at position $\mathbf{r}_0$ is described by

$$
\rho(\mathbf{r}) = q\,\delta^{(3)}(\mathbf{r}-\mathbf{r}_0).
$$

This is the unique compact expression that places all the charge at one point while preserving the total charge under volume integration.

In electrostatics, this charge density appears in Poisson's equation,

$$
\nabla^2 \Phi(\mathbf{r}) = -\frac{\rho(\mathbf{r})}{\varepsilon_0}
= -\frac{q}{\varepsilon_0}\delta^{(3)}(\mathbf{r}-\mathbf{r}_0).
$$

The solution is the familiar Coulomb potential,

$$
\Phi(\mathbf{r}) = \frac{1}{4\pi\varepsilon_0}\frac{q}{|\mathbf{r}-\mathbf{r}_0|}.
$$

### 2. Infinite Line Charge

Consider a line charge of linear density $\lambda$ running along the $z$-axis. Although the charge occupies a one-dimensional set, one can express it as a three-dimensional density by confining it to $x=0$ and $y=0$:

$$
\rho(x,y,z)=\lambda\,\delta(x)\delta(y).
$$

To check the meaning, integrate over a cylinder of finite length $L$ around the axis:

$$
Q = \iiint \rho\,d^3r = \int_{-L/2}^{L/2}\!dz\int dx\int dy\,\lambda\delta(x)\delta(y)=\lambda L.
$$

Thus the volume density correctly reproduces the charge per unit length.

### 3. Infinite Plane Charge

For a plane charge of surface density $\sigma$ lying in the plane $z=0$, the corresponding volume density is

$$
\rho(x,y,z)=\sigma\,\delta(z).
$$

Integrating through a slab of cross-sectional area $A$ and thickness containing $z=0$ gives

$$
Q = \iiint \rho\,d^3r = \iint_A dx\,dy \int dz\,\sigma\delta(z)=\sigma A,
$$

as required for a sheet of charge.

### Why This Simplifies Electrodynamics

Delta functions let us represent point, line, and surface sources using a single formal language of volume densities. That means Gauss's law, Poisson's equation, Green-function methods, and Fourier methods can all be applied without rewriting the theory separately for each geometric type of source.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

# Clean plotting defaults for a notebook-oriented presentation.
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 12
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['axes.titlesize'] = 13
plt.rcParams['legend.fontsize'] = 10


def delta_gaussian_1d(x, eps, center=0.0):
    """Normalized 1D Gaussian approximation to the Dirac delta distribution."""
    return np.exp(-((x - center) / eps) ** 2) / (np.sqrt(np.pi) * eps)


def delta_gaussian_2d(x, y, eps, x0=0.0, y0=0.0):
    """Normalized 2D Gaussian approximation to delta^(2)(r-r0)."""
    return np.exp(-(((x - x0) ** 2 + (y - y0) ** 2) / eps**2)) / (np.pi * eps**2)
```

## Computational Visualizations

The next sections build intuition for the delta distribution using normalized Gaussians. The key idea is that the delta function is approached by a family of ordinary functions whose width shrinks while their total area remains fixed at unity.

Each code cell is preceded by the relevant mathematics and followed by a short discussion of the output.

### Narrow Gaussian Approximations Approaching the Delta Function

We start from the representation

$$
\delta_\varepsilon(x)=\frac{1}{\sqrt{\pi}\,\varepsilon}e^{-x^2/\varepsilon^2}.
$$

As $\varepsilon$ decreases, the profile becomes narrower and its peak grows, while the area remains equal to $1$. This does not converge to an ordinary function; it converges only in the distributional sense.

```python
x = np.linspace(-1.5, 1.5, 4000)
eps_values = [0.5, 0.25, 0.12, 0.06]

fig, ax = plt.subplots()

for eps in eps_values:
    y = delta_gaussian_1d(x, eps)
    ax.plot(x, y, label=fr'$\varepsilon={eps}$')

ax.set_title('Gaussian Approximations to the Dirac Delta')
ax.set_xlabel('$x$')
ax.set_ylabel(r'$\delta_\varepsilon(x)$')
ax.set_xlim(-1.2, 1.2)
ax.legend()
plt.show()
```

The curves become sharply localized near $x=0$ as $\varepsilon$ decreases. The important point is not the increasing height by itself, but the simultaneous narrowing that preserves total area. This is why the Gaussian family is a good approximation to the delta distribution under integration.

### Area Preservation Under the Gaussian Approximation

A defining requirement for any delta-sequence is preservation of total weight:

$$
\int_{-\infty}^{\infty} \delta_\varepsilon(x)\,dx = 1.
$$

We now verify this numerically for several widths by direct quadrature using the trapezoidal rule.

```python
x_area = np.linspace(-4, 4, 20000)
eps_values = np.array([0.8, 0.4, 0.2, 0.1, 0.05])
areas = []

for eps in eps_values:
    y = delta_gaussian_1d(x_area, eps)
    # Numerical approximation to the total area.
    areas.append(np.trapz(y, x_area))

fig, ax = plt.subplots()
ax.plot(eps_values, areas, 'o-', color='tab:blue', label='Numerical area')
ax.axhline(1.0, color='tab:red', linestyle='--', label='Exact area = 1')
ax.set_title('Area Preservation for Gaussian Delta Sequences')
ax.set_xlabel(r'$\varepsilon$')
ax.set_ylabel(r'$\int \delta_\varepsilon(x)\,dx$')
ax.legend()
plt.show()

for eps, area in zip(eps_values, areas):
    print(f'epsilon = {eps:>4.2f}, numerical area = {area:.8f}')
```

The numerical areas stay extremely close to $1$, confirming the central property of the approximation. This is the mechanism by which a sequence of increasingly concentrated functions can still represent a finite amount of charge or probability weight.

### Numerical Illustration of the Sifting Property

To test the sifting property, choose a smooth function such as

$$
f(x)=\cos(2x)+x^2,
$$

and compare the exact value $f(a)$ with the integral

$$
\int_{-\infty}^{\infty} f(x)\delta_\varepsilon(x-a)\,dx.
$$

As $\varepsilon \to 0$, the integral should converge to $f(a)$.

```python
def f(x):
    """Smooth test function used to illustrate the sifting property."""
    return np.cos(2 * x) + x**2


a = 0.6
x_sift = np.linspace(-3, 3, 20000)
eps_values = np.array([0.8, 0.4, 0.2, 0.1, 0.05])
approx_values = []

for eps in eps_values:
    kernel = delta_gaussian_1d(x_sift, eps, center=a)
    # This integral should approach f(a) as epsilon becomes small.
    approx_values.append(np.trapz(f(x_sift) * kernel, x_sift))

exact_value = f(a)

fig, ax = plt.subplots()
ax.plot(eps_values, approx_values, 'o-', label='Integral with Gaussian kernel')
ax.axhline(exact_value, color='tab:red', linestyle='--', label=fr'Exact $f(a)$ at $a={a}$')
ax.set_title('Numerical Sifting Property')
ax.set_xlabel(r'$\varepsilon$')
ax.set_ylabel(r'$\int f(x)\delta_\varepsilon(x-a)\,dx$')
ax.legend()
plt.show()

print(f'Exact f(a) = {exact_value:.8f}')
for eps, value in zip(eps_values, approx_values):
    print(f'epsilon = {eps:>4.2f}, integral = {value:.8f}, error = {abs(value - exact_value):.3e}')
```

The numerical integral approaches $f(a)$ as the Gaussian narrows. This is the computational version of the sifting property. It makes clear that the delta distribution should always be interpreted through what it does inside an integral, not as an ordinary plotted spike.

### Visualization of a Localized 2D Delta Approximation

In two dimensions, the normalized Gaussian approximation is

$$
\delta_\varepsilon^{(2)}(x,y)=\frac{1}{\pi\varepsilon^2}
\exp\left[-\frac{x^2+y^2}{\varepsilon^2}\right].
$$

This has unit integral over the plane and becomes increasingly localized near the origin as $\varepsilon$ decreases. A heatmap is a convenient way to visualize the concentration in the plane.

```python
grid = np.linspace(-1.2, 1.2, 300)
X, Y = np.meshgrid(grid, grid)
eps_2d = 0.18
Z = delta_gaussian_2d(X, Y, eps_2d)

fig, ax = plt.subplots(figsize=(6.5, 5.5))
im = ax.imshow(
    Z,
    extent=[grid.min(), grid.max(), grid.min(), grid.max()],
    origin='lower',
    cmap='inferno',
    aspect='equal'
)
ax.set_title(r'2D Gaussian Approximation to $\delta^{(2)}(\mathbf{r})$')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
fig.colorbar(im, ax=ax, label='Amplitude')
plt.show()

# Numerical check of normalization in 2D.
area_2d = np.trapz(np.trapz(Z, grid, axis=1), grid)
print(f'Approximate 2D integral = {area_2d:.8f}')
```

The heatmap shows strong localization around the origin, with the color scale encoding the large peak near $(0,0)$. The printed integral remains close to $1$, confirming that this sharply peaked object is properly normalized as a two-dimensional delta-sequence.

### 3D Surface Visualization of a Sharply Peaked Approximation

Although the true delta distribution is not an ordinary surface, a 3D rendering of the 2D Gaussian approximation helps visualize the way a localized source becomes simultaneously narrower and taller. This is especially useful when thinking about point-like sources embedded in higher-dimensional space.

```python
grid_surface = np.linspace(-0.8, 0.8, 180)
XS, YS = np.meshgrid(grid_surface, grid_surface)
eps_surface = 0.12
ZS = delta_gaussian_2d(XS, YS, eps_surface)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# A surface plot emphasizes the rapid rise near the localization point.
surf = ax.plot_surface(XS, YS, ZS, cmap='viridis', linewidth=0, antialiased=True)
ax.set_title(r'3D Surface of a Localized Approximation to $\delta^{(2)}(\mathbf{r})$')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('Amplitude')
fig.colorbar(surf, ax=ax, shrink=0.65, pad=0.1)
plt.show()
```

The surface rises very steeply near the origin and falls off rapidly away from it. This conveys the geometric content of a delta-sequence: localization in position space together with normalization of the total integral.

## Summary / Conclusions

The Dirac delta function is best understood as a distribution defined by its action on test functions. It is not an ordinary function, but it behaves in a controlled and highly useful way inside integrals. Its main properties, especially sifting, translation, and scaling, make it an essential tool in graduate electrodynamics.

From the physical side, delta functions provide a compact and rigorous way to encode idealized charge distributions. Point charges, line charges, and surface charges can all be represented within the same volume-density framework used in Maxwell's equations and Poisson's equation. From the computational side, narrow Gaussian sequences provide intuition for how localization and normalization coexist.

This interplay between rigorous distributional language and concrete approximating sequences is exactly what makes the Dirac delta so powerful in theoretical physics.

## Key Takeaways

- The Dirac delta function is a distribution, not an ordinary function.
- Its defining property is the sifting relation $\int f(x)\delta(x-a)\,dx=f(a)$.
- The scaling law $\delta(ax)=\delta(x)/|a|$ preserves normalization under changes of variables.
- In multiple dimensions, the delta distribution factorizes into products such as $\delta^{(3)}(\mathbf{r}-\mathbf{r}_0)=\delta(x-x_0)\delta(y-y_0)\delta(z-z_0)$.
- In electrodynamics, point, line, and surface charges are described compactly by delta-supported volume densities.
- Gaussian delta-sequences give computational intuition: they become narrower and taller while maintaining unit integral.
- Delta functions unify the treatment of localized sources in Maxwell's equations, Poisson's equation, and Green-function methods.
